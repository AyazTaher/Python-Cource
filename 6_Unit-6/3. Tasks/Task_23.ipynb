المهمة 1 – ربط دالتين: استخدام دالة داخل أخرى

المطلوب:

أنشئ دالة multiply(a,b) تعيد حاصل ضرب الرقمين.

أنشئ دالة triple_sum(x,y) تستخدم multiply لحساب (x+y)*3.
</div>

```python
def multiply(a, b):
    return a * b

def triple_sum(x, y):
    return multiply(x + y, 3)

print(triple_sum(4, 5))  # (4+5)*3 = 27
```

المهمة 2 – دالة تتحقق من الرقم وترجع رسالة

المطلوب:

أنشئ دالة باسم check_sign تستقبل رقمًا وتعيد "موجب" إذا الرقم > 0، "سالب" إذا < 0، و "صفر" إذا = 0.
</div>

```python
def check_sign(n):
    if n > 0:
        return "موجب"
    elif n < 0:
        return "سالب"
    else:
        return "صفر"

print(check_sign(-12))
print(check_sign(0))
print(check_sign(7))
```

المهمة 3 – استكشاف النطاق المحلي والعالمي

المطلوب:

أنشئ متغيرًا عالميًا x = 100.

أنشئ دالة local_test() تعرّف متغيرًا محليًا بنفس الاسم x = 50 وتطبعه.
</div>

```python
x = 100  # متغير عالمي

def local_test():
    x = 50  # متغير محلي
    print("داخل الدالة:", x)

local_test()
print("خارج الدالة:", x)
```

المهمة 4 – تعديل متغير عالمي باستخدام global

المطلوب:

أنشئ متغيرًا عالميًا counter = 0.

أنشئ دالة increment_counter() تستخدم global لتعديل قيمة counter بمقدار +1 كل مرة.
</div>

```python
counter = 0

def increment_counter():
    global counter
    counter += 1
    print("داخل الدالة:", counter)

increment_counter()
increment_counter()
increment_counter()
print("خارج الدالة:", counter)
```
