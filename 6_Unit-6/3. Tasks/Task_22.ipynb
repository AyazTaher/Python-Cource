المهمة 1 – دالة جمع الأعداد مع قيمة ابتدائية افتراضية

المطلوب:

أنشئ دالة باسم sum_list تستقبل:

numbers قائمة أعداد

start افتراضي 0

تعيد مجموع الأعداد ابتداءً من start.
</div>

```python
def sum_list(numbers, start=0):
    total = start
    for n in numbers:
        total += n
    return total

print(sum_list([1, 2, 3]))       # 6
print(sum_list([1, 2, 3], 10))   # 16
```

المهمة 2 – دالة تكرار رسالة بعدد مرات افتراضية مع شرط

المطلوب:

أنشئ دالة باسم repeat_message_conditional تستقبل:

msg إلزامي

times افتراضي 3

تطبع الرسالة فقط إذا كانت times أقل من أو تساوي 5، وإلا تطبع "العدد كبير جدًا"
</div>

```python
def repeat_message_conditional(msg, times=3):
    if times > 5:
        print("العدد كبير جدًا")
        return
    for _ in range(times):
        print(msg)

repeat_message_conditional("مرحبا")         # 3 مرات
repeat_message_conditional("Python", 2)     # 2 مرات
repeat_message_conditional("Test", 6)       # العدد كبير جدًا
```

المهمة 3 – دالة خصم السعر مع شرط على النسبة

المطلوب:

أنشئ دالة باسم discount_price_conditional تستقبل:

price إلزامي

rate افتراضي 0.1

إذا كانت rate أكبر من 0.5، تطبع "نسبة الخصم كبيرة جدًا" وتعيد None، وإلا تعيد السعر بعد الخصم.
</div>

```python
def discount_price_conditional(price, rate=0.1):
    if rate > 0.5:
        print("نسبة الخصم كبيرة جدًا")
        return None
    return price * (1 - rate)

print(discount_price_conditional(200))         # 180
print(discount_price_conditional(200, 0.6))    # نسبة الخصم كبيرة جدًا
```

المهمة 4 – دالة تحديد العمر مع معاملات افتراضية وحلقات

المطلوب:

أنشئ دالة باسم check_age تستقبل:

name إلزامي

age افتراضي 0

إذا كان العمر ≥ 18، تطبع "الاسم … بالغ"

إذا كان العمر < 18، تطبع "الاسم … قاصر"
</div>

```python
def check_age(name, age=0):
    if age >= 18:
        print(f"{name} بالغ")
    else:
        print(f"{name} قاصر")

check_age("أحمد", 20)
check_age("سارة", 15)
check_age("ليلى")  # العمر الافتراضي 0 => قاصر
```
