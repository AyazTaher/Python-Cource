المهمة 1 – دالة تطبع جدول الضرب لرقم معين

اكتب دالة باسم multiplication_table(n) تستقبل رقمًا صحيحًا،
ثم تطبع جدول الضرب
</div>

```python
def multiplication_table(n):
    """تطبع جدول الضرب للرقم n."""
    i = 1
    while i <= 10:
        print(f"{n} x {i} = {n * i}")
        i += 1
multiplication_table(3)
```

المهمة 2 – دالة تطبع الأحرف في كلمة

اكتب دالة باسم print_letters(word) تستقبل كلمة،
وتقوم بطباعة كل حرف في سطر مستقل

</div>

```python
def print_letters(word):
    """تطبع كل حرف في الكلمة في سطر مستقل."""
    for ch in word:
        print(ch)
print_letters("Python")
```

المهمة 3 – دالة تعكس نص وتعيده

اكتب دالة باسم reverse_text(text) تستقبل نصًا كسلسلة،
ثم تُعيد النص معكوسًا
</div>

```python
def reverse_text(text):
    """تعيد النص معكوسًا."""
    return text[::-1]
print("reverse_text('Python') =", reverse_text("Python"))
```

المهمة 4 – دالة تتحقق إذا كان الرقم زوجيًا

اكتب دالة باسم is_even(num) تستقبل عددًا صحيحًا كوسيط.

تقوم الدالة بالتحقق مما إذا كان العدد زوجيًا أم لا،

ثم تُعيد القيمة المنطقية True إذا كان زوجيًا، و False إذا كان فرديًا.
</div>

```python
def is_even(num):
    """تتحقق مما إذا كان num زوجيًا."""
    return num % 2 == 0
print("is_even(10) =", is_even(10))
print("is_even(7) =", is_even(7))
```
