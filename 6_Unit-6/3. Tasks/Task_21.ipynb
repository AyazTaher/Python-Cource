المهمة 1 – دالة تجمع قائمة أعداد مع قيمة ابتدائية افتراضية

اكتب دالة باسم sum_numbers(numbers, start=0) تستقبل قائمة أعداد numbers وقيمة ابتدائية start (افتراضية صفر).

تقوم الدالة بجمع الأعداد مع القيمة الابتدائية وإرجاع الناتج.
</div>

المهمة 2 – دالة تطبع رسالة عدة مرات مع قيمة افتراضية

اكتب دالة باسم repeat_message(msg, times=3) تستقبل نص msg وعدد مرات الطباعة times (افتراضي 3)،

وتقوم بطباعة الرسالة عدد المرات المطلوب.
</div>

<div dir="rtl" style="text-align: right;">  

المهمة 3 – دالة تتحقق من طول النص ضد حد افتراضي

اكتب دالة باسم check_length(text, max_length=10) تستقبل نصًا text والحد الأقصى max_length (افتراضي 10).

إذا كان طول النص أكبر من الحد، تطبع "النص طويل جدًا"، وإلا تطبع "النص مقبول".

تستقبل رقمًا number.

تطبع مضاعفات الرقم من 1 حتى times
