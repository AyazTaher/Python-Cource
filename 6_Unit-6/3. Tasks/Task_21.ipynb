المهمة 1 – دالة تجمع قائمة أعداد مع قيمة ابتدائية افتراضية

اكتب دالة باسم sum_numbers(numbers, start=0) تستقبل قائمة أعداد numbers وقيمة ابتدائية start (افتراضية صفر).

تقوم الدالة بجمع الأعداد مع القيمة الابتدائية وإرجاع الناتج.
</div>

```python
def sum_numbers(numbers, start=0):
    """تجمع قائمة أعداد ابتداءً من القيمة start."""
    total = start
    for n in numbers:
        total += n
    return total

print("sum_numbers([1,2,3]) =", sum_numbers([1, 2, 3]))
print("sum_numbers([1,2,3], 10) =", sum_numbers([1, 2, 3], 10))
```

المهمة 2 – دالة تطبع رسالة عدة مرات مع قيمة افتراضية

اكتب دالة باسم repeat_message(msg, times=3) تستقبل نص msg وعدد مرات الطباعة times (افتراضي 3)،

وتقوم بطباعة الرسالة عدد المرات المطلوب.
</div>

```python
def repeat_message(msg, times=3):
    """تطبع الرسالة msg عدد مرات times."""
    count = 0
    while count < times:
        print(msg)
        count += 1

repeat_message("مرحبا")
repeat_message("Python", 2)
```

المهمة 3 – دالة تتحقق من طول النص ضد حد افتراضي

اكتب دالة باسم check_length(text, max_length=10) تستقبل نصًا text والحد الأقصى max_length (افتراضي 10).

إذا كان طول النص أكبر من الحد، تطبع "النص طويل جدًا"، وإلا تطبع "النص مقبول".
</div>

```python
def check_length(text, max_length=10):
    """تتحقق إن كان طول النص <= max_length."""
    if len(text) > max_length:
        print("النص طويل جدًا")
    else:
        print("النص مقبول")

check_length("Hello")
check_length("مرحبا بالعالم")
```

المهمة 4 – دالة تطبع جدول الضرب لعدد محدد مرات افتراضية

اكتب دالة باسم print_multiples(number, times=5) التي:

تستقبل رقمًا number.

تستقبل عدد مرات الطباعة times (افتراضي 5).

تطبع مضاعفات الرقم من 1 حتى times
</div>

```python
def print_multiples(number, times=5):
    """تطبع مضاعفات الرقم number لعدد مرات times."""
    i = 1
    while i <= times:
        print(f"{number} x {i} = {number * i}")
        i += 1

# اختبار الدالة
print_multiples(3)
print("----")
print_multiples(7, 3)
```
