<div dir="rtl" style="text-align: right;">
1-	دالة تحسب مربع رقم وترجع النتيجة</div>

```python

# Function that calculates the square of a number and returns the result
def square(num):
    return num * num

result = square(6)
print(f"The square of 6 is: {result}")
```

<div dir="rtl" style="text-align: right;">
2-	دالة تجمع رقمين وترجع النتيجة</div>

```python

# Function that adds two numbers and returns the result
def add(a, b):
    return a + b

sum_result = add(5, 7)
print(f"The sum of the two numbers is: {sum_result}")
```

<div dir="rtl" style="text-align: right;">
3-	ربط دالتين: استخدام دالة داخل أخرى</div>

```python

# Linking two functions: using one function inside another
def multiply(a, b):
    return a * b

def double_sum(x, y):
    # Using the add function from previous example
    return multiply(add(x, y), 2)

print(double_sum(3, 4))
```

<div dir="rtl" style="text-align: right;">
4-	دالة تتحقق من الرقم وترجع رسالة</div>

```python

# Function that checks a number and returns a message
def check_number(n):
    if n > 0:
        return "Positive"
    elif n < 0:
        return "Negative"
    else:
        return "Zero"
    
print(check_number(-5))
```

 <div dir="rtl" style="text-align: right;">
 مفهوم نطاق المتغيرات (Variable Scope)
</div>

<div dir="rtl" style="text-align: right;">
1-	متغير عالمي
</div>

```python

# Global variable
x = 10  # variable in the global scope

def show_number():
    print(x)  # can access it here

show_number()
print(x)
```

<div dir="rtl" style="text-align: right;">
2-	متغير محلي
</div>

```python

# Local variable
def local_example():
    y = 20  # local variable
    print(y)

local_example()
# Error: y is not defined in the global scope
print(y)
```

<div dir="rtl" style="text-align: right;">
3-	الأولوية بين المحلي والعالمي
</div>

```python

# Priority between local and global variables
z = 100  # global variable

def test_scope():
    # local variable with the same name
    z = 50
    print("Inside the function:", z)

    # Inside the function: 50
    test_scope()
    # Outside the function: 100
print("Outside the function:", z)
```

<div dir="rtl" style="text-align: right;">
4-	استخدام الكلمة المفتاحية global
</div>

```python

# Using the global keyword
a = 5  # global variable

def modify_global():
    # specify that we are referring to the global variable
    global a
    # modify the global variable
    a = 20
print("Inside the function:", a)

# Inside the function: 20
modify_global()
# Outside the function: 20
print("Outside the function:", a)
```
