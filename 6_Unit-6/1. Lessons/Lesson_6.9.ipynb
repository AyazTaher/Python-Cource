```python

# Simple Docstring example
def greet(name):
    """
    This function receives a name and prints a greeting message.
    
    Parameters:
    name (str): The person's name.
    
    Returns:
    None
    """
    print(f"Hello {name}!")
    
    
    # Testing the function with documentation
    # greet("Ahmed")
    # help(greet)
    # """
```

<div dir="rtl" style="text-align: right;">
2-	دالة مع أكثر من معامل</div>

```python

# Function with more than one parameter and Docstring
def add_numbers(a, b):
    """
    This function adds two numbers and returns the result.
    
    Parameters:
    a (int or float): The first number.
    b (int or float): The second number.
    
    Returns:
    int or float: The sum of the two numbers.
    """
    return a + b

# Using the function
print(add_numbers(5, 7))
```

<div dir="rtl" style="text-align: right;">
3-	Docstring متعدد الأسطر</div>

```python

# Multi-line Docstring example
def calculate_area(length, width):
    """
    Calculate the area of a rectangle.
    
    Args:
    length (float): The length of the rectangle.
    width (float): The width of the rectangle.
    
    Returns:
    float: The calculated area.
    
    Note:
    Ensure that the input values are positive.
    """
    return length * width
```

```python

# Function with multiple parameters, print, return, and Docstring
def student_info(name, age, city):
    """
    This function displays information about a student.
    Parameters:
    name (str): The student's name.
    age (int): The student's age.
    city (str): The city where the student lives.
    Returns:
    str: A formatted string containing all the information.
    """
    # Print a welcome message
    print(f"Hello {name}!We are glad to have you with us.")
    #Return a formatted string with the information
    return f"Name: {name}, Age: {age}, City: {city}"
info = student_info("Ahmed", 20, "Damascus")

print(info)
    
```
