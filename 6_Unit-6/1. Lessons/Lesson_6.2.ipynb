الدرس 6.2 – المعاملات (Parameters) والقيم الافتراضية (Default Values)</div>

<div dir="rtl" style="text-align: right;">
1-	 دالة بمعامل إلزامي</div>

```python

def greet(name):
    print(f"Hello, {name}!")

greet("Ahmed")
```

<div dir="rtl" style="text-align: right;">
2-	 دالة بمعامل افتراضي</div>

```python

# A function with a default parameter
def greet(name="Guest"):
    print(f"Hello, {name}!")

# Calling the function without argument
greet()

# Calling the function with an argument
greet("Sarah")
```

<div dir="rtl" style="text-align: right;">
3-	 أكثر من معامل مع خليط بين إلزامي وافتراضي</div>

```python

# A function with multiple parameters, mixing required and default values
def make_profile(name, age=18, country="Not specified"):
    print(f"Name: {name}, Age: {age}, Country: {country}")

# Function calls
make_profile("Ahmed", 25, "Egypt")
make_profile("Sarah", 30)
make_profile("Ali")
```

<div dir="rtl" style="text-align: right;">
4-	 استخدام معاملات متعددة</div>

```python

# Using multiple parameters
def make_shirt(size, text="I love Python"):
    print(f"Shirt size {size} with text: '{text}'")

# Function calls
make_shirt("M")
make_shirt("L", "Fun Programming")
```
