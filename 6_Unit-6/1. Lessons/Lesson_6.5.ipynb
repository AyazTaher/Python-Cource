الدرس 6.5 – المعاملات المتغيرة *args و kwargs</div>

<div dir="rtl" style="text-align: right;">
1-	استخدام args لجمع أعداد متعددة</div>

```python

# Using *args to sum multiple numbers
def sum_all(*numbers):
    total = 0
    for n in numbers:
        total += n
    return total
print(sum_all(1, 2, 3, 4, 5))
```

<div dir="rtl" style="text-align: right;">
2-	استخدام kwargs لعرض معلومات شخص</div>

```python

# Using **kwargs to display a person's information
def print_info(**info):
    for key, value in info.items():
        print(f"{key}: {value}")

print_info(name="Ahmed", age=25, country="Egypt")
```

<div dir="rtl" style="text-align: right;">
3-	دمج *args و kwargs</div>

```python

# Combining *args and **kwargs
def show(*args, **kwargs):
    print("Args:", args)
    print("Kwargs:", kwargs)

show(1, 2, 3, name="Laila", age=30)
```

<div dir="rtl" style="text-align: right;">
4-	حساب متوسط أرقام باستخدام args</div>

```python

# Calculating the average of numbers using *args
def average(*nums):
    return sum(nums) / len(nums)

print(average(10, 20, 30))
```
