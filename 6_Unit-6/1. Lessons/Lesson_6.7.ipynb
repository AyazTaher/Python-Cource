الدرس 6.7 – دوال Lambda أو الدوال المختصرة</div>

<div dir="rtl" style="text-align: right;">
1-	استخدام Lambda لمربع رقم</div>

```python

# Using a Lambda function to square a number
square = lambda x: x**2
print(square(5))
```

<div dir="rtl" style="text-align: right;">
2-	استخدام Lambda جمع رقمين</div>

```python

# Using a Lambda function to add two numbers
add = lambda a, b: a + b
print(add(3, 7))
```

<div dir="rtl" style="text-align: right;">
3-	استخدام Lambda مع قائمة</div>

```python

# Using Lambda with a list
numbers = [1, 2, 3, 4, 5]
squared = list(map(lambda x: x**2, numbers))
print(squared)
```

<div dir="rtl" style="text-align: right;">
4-	استخدام Lambda مع شرط</div>

```python

# Using Lambda with a conditional expression
is_even = lambda n: "even" if n % 2 == 0 else "odd"
print(is_even(7))
```
