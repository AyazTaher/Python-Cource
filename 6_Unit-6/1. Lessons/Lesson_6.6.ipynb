الدرس 6.6 – الدوال المتداخلة Nested Functions</div>

<div dir="rtl" style="text-align: right;">
1-	دالة داخل دالة بسيطة</div>

```python

def outer():
    print("Outer function")
    def inner():
        print("Inner function")
        inner()

outer()
```

<div dir="rtl" style="text-align: right;">
2-	استخدام الدالة الداخلية لمعالجة بيانات</div>

```python

# Using an inner function to process data
def greet(name):
    def format_name():
        return name.upper()
    print(f"Hello {format_name()}!")

greet("ayaz")
```

<div dir="rtl" style="text-align: right;">
3-	دوال متداخلة مع قيمة مرجعة</div>

```python

# Nested functions with a return value
def power(base):
    def square():
        return base ** 2
    return square()

print(power(5))
```

<div dir="rtl" style="text-align: right;">
4-	دالة داخلية تستخدم معاملات متعددة</div>

```python

# Inner functions using multiple arguments
def calculator(x, y):
    def add():
        return x + y
    def multiply():
        return x * y
    return add(), multiply()

print(calculator(3, 4))
```
