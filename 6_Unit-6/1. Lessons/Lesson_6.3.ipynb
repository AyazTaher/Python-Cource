<div dir="rtl" style="text-align: right;">
1-	مثال باستخدام القوائم</div>

```python

# Example using lists
def show_items(items, count=None):
    if count is None:
        # Print all items
        for item in items:
            print(item)
        else:
            # Print only the specified number of items
            for item in items[:count]:
                print(item)

# Data
fruits = ["Apple", "Banana", "Grapes", "Orange", "Mango"]
# Function calls
show_items(fruits)
print("-----")
show_items(fruits, 3)
```

<div dir="rtl" style="text-align: right;">
2-	مثال باستخدام القواميس</div>

```python

# Example using dictionaries
def print_person_info(person, show_country=True):
    print(f"Name: {person['name']}, Age: {person['age']}")
    if show_country:
        print(f"Country: {person['country']}")

# Data
person1 = {"name": "Ahmed", "age": 25, "country": "Egypt"}
person2 = {"name": "Laila", "age": 30, "country": "Jordan"}

# Function calls
print_person_info(person1)          # Prints with country
print("-----")
print_person_info(person2, False)   # Prints without country
```

<div dir="rtl" style="text-align: right;">
3-	مثال عملي على الدمج</div>

```python

# Practical example combining lists, dictionaries, and default parameters
def show_students(students, count=None, show_grade=True):
    selected_students = students if count is None else students[:count]
    for student in selected_students:
        if show_grade:
            print(f"Student: {student['name']}, Grade: {student['grade']}")
        else:
            print(f"Student: {student['name']}")


# Data
students = [
    {"name": "Ahmed", "grade": 90},
    {"name": "Sarah", "grade": 85},
    {"name": "Ali", "grade": 70},
    {"name": "Laila", "grade": 95},
]

# Function calls
# All students with grades
show_students(students)
print("-----")
# First 2 students with grades
show_students(students, 2)
print("-----")
# First 3 students without grades
show_students(students, 3, False)
```
