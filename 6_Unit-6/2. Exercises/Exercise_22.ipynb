التمرين 1 – دالة طباعة جدول الضرب مع حد افتراضي

المطلوب:

أنشئ دالة باسم multiplication_table تستقبل:

number إلزامي

upto افتراضي 10

تطبع جدول الضرب للرقم من 1 حتى upto.
</div>

```python
def multiplication_table(number, upto=10):
    for i in range(1, upto + 1):
        print(f"{number} x {i} = {number * i}")

multiplication_table(3)        # جدول 3 حتى 10
print("---")
multiplication_table(5, 7)     # جدول 5 حتى 7
```

التمرين 2 – دالة التحقق من النصوص الطويلة مع حد افتراضي

المطلوب:

أنشئ دالة باسم check_text_length تستقبل:

text إلزامي

max_length افتراضي 15

إذا كان طول النص أكبر من الحد، تطبع "النص طويل جدًا"، وإلا "النص مقبول".
</div>

```python
def check_text_length(text, max_length=15):
    if len(text) > max_length:
        print("النص طويل جدًا")
    else:
        print("النص مقبول")

check_text_length("مرحبا بكم في بايثون")
check_text_length("Hello")
check_text_length("بايثون", 3)
```
