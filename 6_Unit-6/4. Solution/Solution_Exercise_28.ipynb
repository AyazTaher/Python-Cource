التمرين 1 – دالة ترحيب بسيطة

المطلوب:

اطبع رسالة ترحيب به.

```python

def welcome_message(name):
    """
    This function prints a welcome message for a person.
    
    Parameters:
    name (str): The person's name.
    
    Returns:
    None
    """
    print(f"Hello {name}!Welcome.")
    
# Test the function
welcome_message("Laila")
help(welcome_message)
```

التمرين 2 – دالة جمع أرقام

المطلوب:

أنشئ دالة باسم sum_three_numbers تستقبل ثلاثة أرقام.

ترجع مجموعهم.

أضف Docstring يوضح وظيفة الدالة، نوع المعاملات، والقيمة المرجعة.
</div>

```python

def sum_three_numbers(a, b, c):
    """
    This function adds three numbers and returns the result.
    
    Parameters:
    a (int or float): The first number.
    b (int or float): The second number.
    c (int or float): The third number.
    
    Returns:
    int or float: The sum of the three numbers.
    """
    return a + b + c

# Test the function
result = sum_three_numbers(5, 10, 3)
print("The sum is:", result)
help(sum_three_numbers)
```
