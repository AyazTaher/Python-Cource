```python

def calculate_rectangle_area(length, width):
    """
    Calculate the area of a rectangle.
    
    Parameters:
    length (float): The length of the rectangle.
    width (float): The width of the rectangle.
    
    Returns:
    float: The calculated area.
    
    Note:
    Make sure that length and width are positive.
    """
    return length * width

# Test the function
area = calculate_rectangle_area(5, 3)
print("Rectangle area:", area)
help(calculate_rectangle_area)
```

```python

def employee_info(name, age, position, salary):
    """
    This function displays employee information and returns a formatted text.
    
    Parameters:
    name (str): Employee's name.
    age (int): Employee's age.
    position (str): Job title or position.
    salary (float): Salary amount.
    
    Returns:
    str: A formatted string containing all the information.
    """
    print(f"Hello {name}!We welcome you to the team.")
    return f"Name: {name}, Age: {age}, Position: {position}, Salary: {salary}"

# Test the function
info = employee_info("Ahmed", 30, "Engineer", 1200.50)
print(info)
help(employee_info)
```
