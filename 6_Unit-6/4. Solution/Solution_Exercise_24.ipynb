التمرين 1 – جمع أعداد متعددة باستخدام args

المطلوب:

أنشئ دالة باسم sum_numbers تستخدم *args لجمع أي عدد من الأرقام.

اطبع ناتج جمع 3 أرقام، ثم 5 أرقام.
</div>

```python

def sum_numbers(*numbers):
    total = 0
    for n in numbers:
        total += n
    return total
    
print(sum_numbers(5, 10, 15))
print(sum_numbers(1, 2, 3, 4, 5))


```

*التمرين 2 – عرض معلومات شخص باستخدام kwargs

المطلوب:

أنشئ دالة باسم display_info تستخدم **kwargs لعرض أي معلومات مسماة عن شخص.
</div>

```python

def display_info(**info):
    for key, value in info.items():
        print(f"{key}: {value}")

display_info(name="Ahmed", age=25, country="Egypt")
```
