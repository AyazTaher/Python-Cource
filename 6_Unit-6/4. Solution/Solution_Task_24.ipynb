المهمة 1 – دمج *args و kwargs

المطلوب:

أنشئ دالة باسم show_all تستقبل:

*args لعناصر غير مسماة

**kwargs لعناصر مسماة

اطبع كلاهما بشكل واضح.
</div>

```python

def show_all(*args, **kwargs):
    print("Args:", args)
    print("Kwargs:", kwargs)

show_all(1, 2, 3, name="Laila", age=30)


```

المهمة 2 – حساب المتوسط باستخدام args

المطلوب:

أنشئ دالة باسم average_numbers تحسب متوسط أي عدد من الأرقام باستخدام *args.
</div>

```python

def average_numbers(*nums):
    return sum(nums)/len(nums)

print(average_numbers(10, 20, 30, 40))
print(average_numbers(5, 15, 25, 35, 45, 55))
```

المهمة 3 – دمج *args و kwargs لحساب العمليات الرياضية

المطلوب:

أنشئ دالة calculate تستقبل:

*args أرقام

**kwargs مع مفتاح operation لتحديد العملية: "sum" للجمع أو "product" للضرب.
</div>

```python

def calculate(*args, **kwargs):
    op = kwargs.get("operation", "sum")
    if op == "sum":
        result = sum(args)
    elif op == "product":
        result = 1
        for n in args:
            result *= n
    else:
        return "Unknown operation"
    return result
    
print(calculate(2, 3, 4, operation="sum"))      # 9
print(calculate(2, 3, 4, operation="product"))  # 24
```

المهمة 4 – إنشاء تقرير ديناميكي باستخدام *args و kwargs

المطلوب:

أنشئ دالة report تستقبل:

*args لعنوان التقرير أو عناصر أساسية

**kwargs لمعلومات إضافية (مثل التاريخ، الكاتب، القسم)
</div>

```python

def report(*args, **kwargs):
    print("Report Title / Key Items:")
    for item in args:
        print("-", item)
        print("Additional Information:")
        for key, value in kwargs.items():
            print(f"{key}: {value}")

report("Q1 Sales", "Total Revenue", name="Ahmed", date="2025-10-30", department="Sales")
```
