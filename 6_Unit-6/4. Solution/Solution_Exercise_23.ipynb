التمرين 1 – دالة تحسب مربع الرقم وترجع النتيجة

المطلوب:

أنشئ دالة باسم square_number تستقبل رقمًا وتعيد مربعه باستخدام return.

```python

def square_number(num):
    return num * num

result = square_number(8)
print(f"The square of 8 is: {result}")


```

التمرين 2 – دالة تجمع رقمين وترجع النتيجة

المطلوب:

أنشئ دالة باسم sum_two تستقبل رقمين وتعيد مجموعهما.
</div>

```python

def sum_two(a, b):
    return a + b

total = sum_two(10, 15)
double_total = total * 2
print(f"Sum: {total}, Double: {double_total}")
```
