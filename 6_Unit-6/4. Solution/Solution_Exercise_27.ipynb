التمرين 1 – تربيع قائمة أعداد باستخدام map

استخدم Lambda مع map() لتربيع جميع عناصر القائمة [1,2,3,4,5].
</div>

```python

numbers = [1,2,3,4,5]
squared = list(map(lambda x: x**2, numbers))
print(squared)
```

<div dir="rtl" style="text-align: right;">  

التمرين 2 – فلترة الأعداد الزوجية باستخدام filter

استخدم Lambda مع filter() لاستخراج الأعداد الزوجية فقط من [10,15,20,25,30].
</div>

```python

nums = [10,15,20,25,30]
evens = list(filter(lambda x: x%2==0, nums))
print(evens)
```

التمرين 3 – ترتيب قائمة من الأعداد تنازليًا

استخدم Lambda مع sorted() لترتيب [5,2,9,1,7] تنازليًا.
</div>

```python

nums = [5,2,9,1,7]
sorted_nums = sorted(nums, key=lambda x: -x)
print(sorted_nums)
```

<div dir="rtl" style="text-align: right;">  

 التمرين 4 – تحويل قائمة نصوص إلى أحرف كبيرة

استخدم Lambda مع map() لتحويل ["ahmed","sara","ali"] إلى أحرف كبيرة. النص

</div>

```python

names = ["ahmed","sara","ali"]
upper_names = list(map(lambda x: x.upper(), names))
print(upper_names)
```

التمرين 5 – فلترة النصوص حسب الطول

استخدم Lambda مع filter() لاستخراج النصوص التي طولها > 4 من ["Python","C","Java","Go"].
</div>

```python

words = ["Python","C","Java","Go"]
long_words = list(filter(lambda w: len(w)>4, words))
print(long_words)

```

التمرين 6 – جمع قائمة من القيم باستخدام reduce

استخدم Lambda مع functools.reduce() لجمع [1,2,3,4,5].
</div>

```python

from functools import reduce
nums = [1,2,3,4,5]
total = reduce(lambda a,b: a+b, nums)
print(total)

```
