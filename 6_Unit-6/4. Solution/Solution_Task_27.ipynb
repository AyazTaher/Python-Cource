المهمة 1 – فرز قائمة من القواميس حسب العمر

لديك:
people = [{"name":"أحمد","age":25},{"name":"ليلى","age":30},{"name":"علي","age":20}]

استخدم Lambda مع sorted() لفرز القائمة حسب العمر تصاعديًا.

</div>

```python

people = [{"name": "Ahmed", "age": 25},
          {"name": "Laila", "age": 30},
          {"name": "Ali", "age": 20}]

sorted_people = sorted(people, key=lambda p: p["age"])
print(sorted_people)
```

<div dir="rtl" style="text-align: right;">  

المهمة 2 – ترتيب القوائم بناءً على الحرف الأخير

المطلوب: ["sun","moon","star","sky"] → ترتيب حسب الحرف الأخير
</div>

```python

words = ["sun","moon","star","sky"]
sorted_words = sorted(words, key=lambda x: x[-1])
print(sorted_words)

```

المهمة 3 – دمج map و sorted

المطلوب: [5,2,9,1,7] → مضاعفة الأرقام ثم ترتيب تصاعدي
</div>

```python

numbers = [5, 2, 9, 1, 7]
result = sorted(map(lambda x: x*2, numbers))
print(result)

```

المهمة 5 – تحويل درجات الطلاب إلى درجات بالحروف

الوصف: لدينا قائمة درجات [95, 82, 67, 58, 74] ونريد تحويلها إلى درجات بالحروف (A, B, C, D, F) باستخدام map و lambda
</div>

```python

grades = [95, 82, 67, 58, 74]

letter_grades = list(map(lambda x:
                         "A" if x >= 90 else
                         "B" if x >= 80 else
                         "C" if x >= 70 else
                         "D" if x >= 60 else
                         "F", grades))

print(letter_grades)
```

المهمة 6 – اختيار الأسماء التي تبدأ بحرف معين

الوصف: لدينا قائمة أسماء ["Ahmed","Sara","Ali","Fatima","Omar"] ونريد استخراج الأسماء التي تبدأ بحرف 'A' باستخدام filter و lambda
</div>

```python

names = ["Ahmed","Sara","Ali","Fatima","Omar"]

a_names = list(filter(lambda x: x.startswith("A"), names))
print(a_names)
```
