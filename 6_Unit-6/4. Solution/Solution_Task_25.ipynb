المهمة 1 – دالة داخلية مع قيمة مرجعة

المطلوب:

أنشئ دالة باسم cube(base) تقوم بإرجاع مكعب الرقم باستخدام دالة داخلية calc_cube().

اطبع نتيجة المكعب.
</div>

```python

def cube(base):
    def calc_cube():
        return base ** 3
    return calc_cube()

print(cube(4))
```

المهمة 2 – دوال متداخلة لحساب عمليات رياضية متعددة

المطلوب:

أنشئ دالة calculator(x, y) تحتوي على:

دالة داخلية add() لإرجاع مجموع x و y

دالة داخلية multiply() لإرجاع حاصل ضرب x و y

ارجع كلا الناتجين من الدالة الخارجية.
</div>

```python

def calculator(x, y):
    def add():
        return x + y
    def multiply():
        return x * y
    return add(), multiply()

print(calculator(3, 4))
```

المهمة 3 – دالة داخلية للتحقق من طول النص

المطلوب:

أنشئ دالة باسم check_text(text, max_length) تحتوي على دالة داخلية is_valid() تتحقق إذا كان طول النص أقل أو يساوي max_length.

ارجع "مقبول" إذا النص صالح و "طويل جدًا" إذا النص أطول.
</div>

```python

def check_text(text, max_length):
    def is_valid():
        return len(text) <= max_length
    return "Acceptable" if is_valid() else "Too long"

print(check_text("Hello", 10))
print(check_text("Hello World ", 10))
```

المهمة 4 – دالة داخلية تتعامل مع قائمة

المطلوب:

أنشئ دالة process_numbers(numbers) تستقبل قائمة أرقام.

داخلها دالة square_numbers() تعيد قائمة جديدة بمربعات الأرقام.

اطبع النتيجة بعد استدعاء الدالة الخارجية.
</div>

```python

def process_numbers(numbers):
    def square_numbers():
        return [n**2 for n in numbers]
    return square_numbers()

nums = [1, 2, 3, 4]
print(process_numbers(nums))
```
