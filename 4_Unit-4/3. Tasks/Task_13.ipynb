المهمة 1 – تصنيف الموظف حسب العمر والخبرة

المطلوب:

اطلب من المستخدم إدخال العمر وعدد سنوات الخبرة، ثم:
1.	إذا كان العمر < 25 → اطبع "مبتدئ".
2.	إذا كان العمر بين 25 و40:
-	إذا كانت الخبرة ≥ 5 → اطبع "متوسط".
-	وإلا → اطبع "مبتدئ".
3.	إذا كان العمر > 40 → اطبع "خبير" دائمًا.

المهمة 2 – تحديد خصم المتجر

المطلوب:

اطلب من المستخدم إدخال نوع العميل ("regular" أو "vip") وقيمة الشراء، ثم:
1.	إذا كان العميل VIP:
-	إذا كانت القيمة ≥ 500 → اطبع "خصم 20%".
-	وإلا → اطبع "خصم 10%".
2.	إذا كان العميل عادي:
-	إذا كانت القيمة ≥ 500 → اطبع "خصم 5%".
- وإلا → اطبع "لا يوجد خصم".
