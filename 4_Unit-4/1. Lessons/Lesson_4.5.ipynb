البيانات </div>

```python
data = [
    {"Age": 23, "Sex": "F", "BP": "HIGH", "Cholesterol": "HIGH", "Drug": "drugY"},
    {"Age": 47, "Sex": "M", "BP": "LOW", "Cholesterol": "HIGH", "Drug": "drugC"},
    {"Age": 28, "Sex": "F", "BP": "NORMAL", "Cholesterol": "HIGH", "Drug": "drugX"},
    {"Age": 61, "Sex": "F", "BP": "LOW", "Cholesterol": "HIGH", "Drug": "drugY"},
    {"Age": 22, "Sex": "F", "BP": "NORMAL", "Cholesterol": "HIGH", "Drug": "drugX"}
]
print(data)
```

<div dir="rtl" style="text-align: right;">
تمرين 1: تصنيف ضغط الدم
المطلوب: طباعة العمر ومستوى ضغط الدم لكل مريض كـ: "عالي، متوسط، منخفض"</div>

```python

for patient in data:
    if patient["BP"] == "HIGH":
        level = "High"
    elif patient["BP"] == "NORMAL":
        level = "Normal"
    else:
        level = "Low"
        print(patient["Age"], "years old - Blood Pressure:", level)
        
```

<div dir="rtl" style="text-align: right;">
تمرين 2: تحديد مستوى الخطر باستخدام الشروط المتداخلة
المطلوب: بناءً على ضغط الدم والكوليسترول، اطبع مستوى الخطر لكل مريض: "خطر شديد، خطر متوسط، خطر منخفض"</div>

```python

for patient in data:
    risk = (
        "High Risk" if patient["BP"] == "HIGH" and patient["Cholesterol"] == "HIGH"
        else "Moderate Risk" if patient["BP"] == "HIGH" or patient["Cholesterol"] == "HIGH"
        else "Low Risk")
    print(patient["Age"], "years old - Risk Level:", risk)


```

<div dir="rtl" style="text-align: right;">
تمرين 3: استخدام continue و break
المطلوب: طباعة أول مريضين يستخدمان drugY، وتخطي الآخرين</div>

```python

count = 0
for patient in data:
    if patient["Drug"]!= "drugY":
        continue
    print(patient["Age"], "years old - Drug:", patient["Drug"])
    count += 1
    if count == 2:
        break


```
