<div dir="rtl" style="text-align: right;">
2-	 التحقق من القبول في الوظيفة</div>

```python

age = 17
experience_years = 1

if age >= 18:
    if experience_years >= 2:
        print("Eligible to apply for the job")
    else:
        print("Age issu itable but experience is insufficient")
else:
    print("Not eligible due to age")
```

<div dir="rtl" style="text-align: right;">
3-	 تسجيل الدخول مع التحقق من الصلاحيات</div>

```python
username = "admin"
password = "1234"

if username == "admin":
    if password == "1234":
        print("Login successful  you have admin privileges")
    else:
        print("Username is correct but password is wrong")
else:
    print("Username does not exist")

```

<div dir="rtl" style="text-align: right;">
4-	 تحديد حالة الطقس</div>

```python

temperature = 20
humidity = 85

if temperature > 25:
    if humidity > 80:
        print("The weather is hot and humid")
    else:
        print("The weather is hot but humidity is low")
else:
    print("The weather is mild or cold")
```
