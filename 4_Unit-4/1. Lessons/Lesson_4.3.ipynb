<div dir="rtl" style="text-align: right;">
1-	 التحقق من القيم في الشرط</div>

```python

value = 4

if value:
    print("The value is considered True")
else:
    print("The value is considered False")
```

<div dir="rtl" style="text-align: right;">
2-	 التحقق من قائمة فارغة</div>

```python

my_list = ["1","2"]

if my_list:
    print("The list contains elements")
else:
    print("The list is empty")
```

<div dir="rtl" style="text-align: right;">
3-	 التحقق من النصوص</div>

```python

name = "Dan"

if name:
    print("A name has been entered")
else:
    print("No name has been entered")
```

المطلوب:
اكتب برنامجًا يطلب من المستخدم إدخال رقم، ثم:

1.	إذا كانت القيمة المدخلة فارغة (لم يدخل شيء) → اطبع "لم تدخل أي قيمة"
2.	إذا كانت القيمة المدخلة تساوي صفرًا → اطبع "أدخلت صفر"
3.	إذا كانت القيمة غير صفرية → اطبع "أدخلت رقمًا صالحًا"

```python

value = input("Enter a number: ")

if not value:
    print("No value was entered")
else:
    number = int(value)
    if number == 0:
        print("You entered zero")
    else:
        print("You entered a valid number")
        
```
