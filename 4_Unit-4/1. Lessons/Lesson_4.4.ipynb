الدرس 4.3 – اختصار الجمل الشرطية (Simplifying Conditions)</div>

<div dir="rtl" style="text-align: right;">
1-	الجمع بين الشروط باستخدام and و or</div>

```python

age = 15
experience = 3

if age >= 18 and experience >= 2:
    print("Eligible to apply")
else:
    print("Not eligible")

    ###
day = "Monday"

if day == "Friday" or day == "Saturday":
    print("Holiday")
else:
    print("Workday")
```

<div dir="rtl" style="text-align: right;">
2-	الجملة الشرطية المختصرة (Ternary Operator)</div>

```python

score = 42
result = "Pass" if score >= 50 else "Fail"
print(result)
```

<div dir="rtl" style="text-align: right;">
3-	استخدام in و not in</div>

```python


if day == "Friday" or day == "Saturday" or day == "Sunday":
    print("Holiday")
else:
    print("Workday")

    ##
if day in ["Friday", "Saturday", "Sunday"]:
    print("Holiday")
else:
    print("Condition not met")

        ##
if day not in ["Friday", "Saturday"]:
    print("Workday")
else:
    print("Holiday")
```

<div dir="rtl" style="text-align: right;">
4-	استخدام or و and للقيم الافتراضية</div>

```python

name = input("Enter your name: ") or "Guest"
print("Hello", name)

##

x = 5
y = 10

max_value = x if x > y else y
print(max_value)
```

<div dir="rtl" style="text-align: right;">
5-	استخدام Dictionary بدل if متعددة</div>

```python

status = "B"

grades = {
    "A": "Excellent",
    "B": "Good",
    "C": "Average"
}

print(grades.get(status, "Fail"))
```

<div dir="rtl" style="text-align: right;">
6-	طريقة سلسلة المقارنات (Operator Chaining)</div>

```python

x = 8
if x > 1 and x < 10:
    print("x is within the range")
else:
    print("Condition")

    ###
x = 10

if 1 < x < 10:
        print("x is within the range")
else:
        print("Condition")

```

<div dir="rtl" style="text-align: right;">
-  التحقق من النطاق</div>

```python

x = 11

if 1 < x < 10:
    print("x is between 1 and 10")
else:
    print("Condition")
```

<div dir="rtl" style="text-align: right;">
- أكثر من مقارنة في سلسلة واحدة</div>

```python

y = 20

if 10 < y < 20 < 30:
    print("y is between 10 and 20, and 20 is also less than 30")
else:
    print("Condition")
```

<div dir="rtl" style="text-align: right;">
-	الجمع بين المساواة والمقارنة</div>

```python

age = 35

if 18 <= age < 30:
    print("Age is suitable for joining")
else:
    print("Condition")
```

<div dir="rtl" style="text-align: right;">
-	استخدام == و != في السلسلة</div>

```python

z = 12

if 1 < z!= 3 < 10:
    print("z is not 3, and it is also between 1 and 10")
else:
    print("Condition")
```

<div dir="rtl" style="text-align: right;">
-	مقارنة النصوص باستخدام operator chaining</div>

```python

word = "mango"

if "apple" < word < "zebra":
    print("The word is alphabetically between apple and zebra")
else:
    print("Condition")
    
```

الطريقة التقليدية:</div>

```python

day = input("Enter the day: ")

if day == "Friday" or day == "Saturday":
    print("Holiday")
elif day == "Sunday":
    print("Half day")
else:
    print("Workday")
    
```

<div dir="rtl" style="text-align: right;">
الطريقة المختصرة باستخدام Dictionary:</div>

```python

day = input("Enter the day: ")

schedule = {
    "Friday": "Holiday",
    "Saturday": "Holiday",
    "Sunday": "Half day"
}

print(schedule.get(day, "Workday"))
```

<div dir="rtl" style="text-align: right;">
باستخدم operator chaining</div>

```python

score = int(input("Enter the exam score (0 to 100): "))

# Checking using operator chaining
if 90 <= score <= 100:
    print("Excellent")
elif 75 <= score <= 89:
    print("Very Good")
elif 50 <= score <= 74:
    print("Pass")
elif 0 <= score < 50:
    print("Fail")
else:
    print("Invalid score, please enter a value between 0 and 100")
    
```
