<div dir="rtl" style="text-align: right;">
هنالك عدة ملاحظات أساسية حول جمل الشرط:</div>

<div dir="rtl" style="text-align: right;">

-	يجب أن ينتهي السطر الذي يحتوي على if / elif / else بنقطتين رأسيتين 
-	المسافة البادئة (Indentation) مهمة جدًا في بايثون، وهي التي تحدد الكود التابع لكل شرط
-	يمكن أن تحتوي جملة if على شرط واحد، أو أن تكون سلسلة من if / elif / else
-	القيم التي تعتبر True و False تعتمد على مفهوم Truthy / Falsy الذي سنتطرق له في الدرس 4.3
</div>

<div dir="rtl" style="text-align: right;">
2-	 شرط بسيط باستخدام if</div>

```python

age = 17
if age >= 18:
    print("You are an adult and can get a driving license")
else:
    print("You cannot get a driving license")

```

<div dir="rtl" style="text-align: right;">
3-	 استخدام if و else</div>

```python


temperature = 36

if temperature >= 35:
    print("The weather is hot today")
else:
    print("The weather is mild or cold today")
```

<div dir="rtl" style="text-align: right;">
4-	 استخدام if و elif و else</div>

```python

score = 58

if score >= 90:
    print("Excellent")
elif score >= 75:
    print("Very Good")
elif score >= 60:
    print("Good")
else:
    print("Fail")
```
