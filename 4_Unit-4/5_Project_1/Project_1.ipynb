```python
month1 = {
    "Month": "1/2025",
    "Groceries": 7892,
    "Rent": 13000,
    "Transportation": 2419,
    "Gym": 891,
    "Utilities": 2238,
    "Healthcare": 1956,
    "Investments": 7590,
    "Savings": 11369,
    "EMI/Loans": 6000,
    "Dining & Entertainment": 2622,
    "Shopping & Wants": 2571,
    "Total Expenditure": 47179,
    "Income": 64000
}

month2 = {
    "Month": "2/2025",
    "Groceries": 7687,
    "Rent": 13000,
    "Transportation": 2713,
    "Gym": 923,
    "Utilities": 1536,
    "Healthcare": 2162,
    "Investments": 8814,
    "Savings": 7059,
    "EMI/Loans": 0,
    "Dining & Entertainment": 3048,
    "Shopping & Wants": 1278,
    "Total Expenditure": 41161,
    "Income": 64000
}

print(month1)
print(month2)
```

الطلب 1 – حساب الفائض الشهري

 حساب الفائض الشهري بناءً على الدخل والمصروفات الشهرية، ثم طباعته.

</div>

```python
print("=== التمرين 1: حساب الفائض الشهري ===")

income = month1["Income"]
total_expense = month1["Total Expenditure"]
surplus = income - total_expense
print(f"Month: {month1['Month']}, Income: {income}, Expenditure: {total_expense}, Surplus: {surplus}")
```

<div dir="rtl" style="text-align: right;">  

الطلب 2 – التحقق من الفائض أو العجز

تحقق مما إذا كان الفائض موجبًا، صفرًا، أم سالبًا، واطبع الحالة.


</div>

```python
print("=== التمرين 2: تحقق من الفائض أو العجز ===")

if surplus > 0:
    print("Surplus")
elif surplus == 0:
    print("Balanced")
    
```

<div dir="rtl" style="text-align: right;">  

الطلب 3 – حساب نسبة المصروف لكل بند

احسب نسبة مصروف Groceries وRent وTransportation من إجمالي المصروفات.

</div>

```python
print("=== التمرين 3: نسبة كل مصروف من المجموع ===")

groceries_percent = (month1["Groceries"] / total_expense) * 100
rent_percent = (month1["Rent"] / total_expense) * 100
transport_percent = (month1["Transportation"] / total_expense) * 100

print(f"Groceries: {groceries_percent:.2f}%")

print(f"Rent: {rent_percent:.2f}%")

print(f"Transportation: {transport_percent:.2f}%")
```
