المهمة 1 – تحديد أكبر رقم بين رقمين


المطلوب:

اطلب من المستخدم إدخال رقمين، ثم:
- 	اطبع الرقم الأكبر باستخدام مشروط ثلاثي.

الحل:



</div>

```python

a = float(input("Enter the first number: "))
b = float(input("Enter the second number: "))
print(a if a > b else b)
```

المهمة 2 – تحديد حالة الطالب

المطلوب:

اطلب من المستخدم إدخال درجة الطالب (0-100)، ثم:
-	"ناجح" إذا الدرجة ≥ 50، وإلا "راسب"
-	استخدم اختصار الجملة الشرطية بدلاً من if التقليدي.

الحل:



</div>

```python

score = float(input("Enter the student's score: "))
result = "Pass" if score >= 50 else "Fail"
print(result)

```
