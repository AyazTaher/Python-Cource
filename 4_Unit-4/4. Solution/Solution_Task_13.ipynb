المهمة 1 – تصنيف الموظف حسب العمر والخبرة

المطلوب:

اطلب من المستخدم إدخال العمر وعدد سنوات الخبرة، ثم:
1.	إذا كان العمر < 25 → اطبع "مبتدئ".
2.	إذا كان العمر بين 25 و40:
-	إذا كانت الخبرة ≥ 5 → اطبع "متوسط".
-	وإلا → اطبع "مبتدئ".
3.	إذا كان العمر > 40 → اطبع "خبير" دائمًا.
الحل:



</div>

```python

age = int(input("Enter your age:" ))
experience = int(input("Enter your years of experience: "))

if age < 25:
    print("Beginner")
else:
    if 25 <= age <= 40:
        if experience >= 5:
            print("Intermediate")
        else:
            print("Beginner")
    else:
        print("Expert")
```

المهمة 2 – تحديد خصم المتجر

المطلوب:

اطلب من المستخدم إدخال نوع العميل ("regular" أو "vip") وقيمة الشراء، ثم:
1.	إذا كان العميل VIP:
-	إذا كانت القيمة ≥ 500 → اطبع "خصم 20%".
-	وإلا → اطبع "خصم 10%".
2.	إذا كان العميل عادي:
-	إذا كانت القيمة ≥ 500 → اطبع "خصم 5%".
- وإلا → اطبع "لا يوجد خصم".
الحل:



</div>

```python
customer_type = input("Enter customer type (regular/vip): ")
purchase = float(input("Enter purchase amount: "))

if customer_type == "vip":
    if purchase >= 500:
        print("20% discount")
    else:
        print("10% discount")
else:
    if purchase >= 500:
        print("5% discount")
    else:
        print("No discount")
```
