التمرين 1 – تصنيف العمر
المطلوب:

اطلب من المستخدم إدخال العمر، ثم:
1.	إذا كان العمر < 5 → اطبع "رضيع".
2.	إذا كان العمر بين 5 و12 → اطبع "طفل صغير".
3.	إذا كان العمر بين 13 و17 → اطبع "مراهق".
4.	إذا كان العمر ≥ 18 → اطبع "بالغ".

```python

age = int(input("Enter your age: "))

if age < 13:
    print("Child")
else:
    if age <= 19:
        print("Teenager")
    else:
        if age <= 59:
            print("Adult")
        else:
            print("Elderly")


            
```

الطريقة الثانية
</div>

```python

age = int(input("Enter your age: "))

if age < 13:
    print("Child")
elif 13 <= age <= 19:
    print("Teenager")
elif 20 <= age <= 59:
    print("Adult")
else:
    print("Elderly")


    
```
