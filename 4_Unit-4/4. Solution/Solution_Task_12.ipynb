المهمة 1 – تصنيف العمر مع فئات إضافية
المطلوب:

اطلب من المستخدم إدخال العمر، ثم:
1.	إذا كان العمر ≤ 3 → اطبع "رضيع جدًا".
2.	إذا كان العمر بين 4 و10 → اطبع "طفل صغير".
3.	إذا كان العمر بين 11 و18 → اطبع "مراهق".
4.	إذا كان العمر بين 19 و50 → اطبع "بالغ".
5.	إذا كان العمر ≥ 51 → اطبع "كبير في السن".
الحل:



</div>

<div dir="rtl" style="text-align: right;">  

الطريقة الأولى
</div>

```python

age = int(input("Enter your age: "))

if age <= 3:
    print("Very infant")
else:
    if age <= 10:
        print("Young child")
    else:
        if age <= 18:
            print("Teenager")
        else:
            if age <= 50:
                print("Adult")
            else:
                print("Elderly")
```

المهمة 2 – تصنيف العمر حسب مراحل الحياة
المطلوب:

اطلب من المستخدم إدخال العمر، ثم:
1.	إذا كان العمر < 6 → اطبع "طفل صغير".
2.	إذا كان العمر بين 6 و12 → اطبع "طفل كبير".
3.	إذا كان العمر بين 13 و18 → اطبع "مراهق".
4.	إذا كان العمر بين 19 و35 → اطبع "شاب".
5.	إذا كان العمر بين 36 و60 → اطبع "بالغ".
6.	إذا كان العمر > 60 → اطبع "كبير في السن".
الحل:



</div>

<div dir="rtl" style="text-align: right;">  

الطريقة الأولى
</div>

```python

age = int(input("Enter your age: "))

if age < 6:
    print("Young child")
else:
    if age <= 12:
        print("Older child")
    else:
        if age <= 18:
            print("Teenager")
        else:
            if age <= 35:
                print("Young adult")
            else:
                if age <= 60:
                    print("Adult")
                else:
                    print("Elderly")

                    
```
