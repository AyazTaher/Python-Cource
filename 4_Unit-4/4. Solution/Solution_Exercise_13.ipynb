التمرين 1 – تقييم الطالب حسب الدرجة

المطلوب:

اطلب من المستخدم إدخال درجة الطالب (0-100)، ثم:
1.	إذا كانت الدرجة ≥ 90 → اطبع "امتياز".
2.	إذا كانت الدرجة ≥ 75 وأقل من 90 → اطبع "جيد جدًا".
3.	إذا كانت الدرجة ≥ 60 وأقل من 75 → اطبع "جيد".
4.	إذا كانت الدرجة أقل من 60 → اطبع "راسب".
الحل:


</div>

```python

score = int(input("Enter the student's score (0100): "))

if score >= 90:
    print("Excellent")
elif 75 <= score < 90:
    print("Very Good")
elif 60 <= score < 75:
    print("Good")
else:
    print("Fail")
```

التمرين 2 – تحديد تذكرة دخول الحديقة

المطلوب:

اطلب من المستخدم إدخال العمر ويوم الأسبوع (مثلاً: "weekend" أو "weekday")، ثم:
1.	إذا كان العمر < 5 → اطبع "دخول مجاني".
2.	إذا كان العمر بين 5 و17:
- 	إذا كان يوم العطلة → اطبع "تذكرة نصف السعر".
- 	وإلا → اطبع "تذكرة كاملة".
3.	إذا كان العمر ≥ 18 → اطبع "تذكرة كاملة" دائمًا.
الحل:


</div>

<div dir="rtl" style="text-align: right;">  


</div>

```python

age = int(input("Enter your age: "))
day = input("Enter the day type (weekday or weekend): ").lower()

if age < 5:
    print("Free entry")
elif 5 <= age <= 17:
    if day == "weekend":
        print("Half-price ticket")
    else:
        print("Full-price ticket")
else:
        print("Full-price ticket")

```
