المهمة 1 – كلمة السر بثلاث محاولات

المطلوب:

1.	الكلمة الصحيحة هي "python123".
2.	اسمح للمستخدم بـ 3 محاولات فقط.
3.	إذا كانت صحيحة، اطبع "تم تسجيل الدخول بنجاح!".
4.	إذا أخطأ في الثلاث مرات، اطبع "تم حظر الحساب!".


</div>

```python

# التحقق من كلمة السر مع 3 محاولات فقط

password = "python123"
tries = 0

while tries < 3:
    user_input = input("أدخل كلمة السر: ")
    if user_input == password:
        print("تم تسجيل الدخول بنجاح!")
        break
    else:
        print("كلمة السر غير صحيحة! حاول مرة أخرى.")
        tries += 1

if tries == 3:
    print("تم حظر الحساب!")
```

المهمة 2 – آلة حاسبة تفاعلية
المطلوب:

1.	أنشئ قائمة خيارات:
2. جمع
3. طرح
4. ضرب
5. قسمة
6. خروج
-	اطلب من المستخدم اختيار رقم العملية.
-	أدخل رقمين، ثم نفّذ العملية المختارة.
-	استمر في الحلقة حتى يختار المستخدم رقم (5).



</div>

```python

# آلة حاسبة بسيطة باستخدام while loop

choice = 0

while choice != 5:
    print("\nاختر عملية:")
    print("1 - جمع")
    print("2 - طرح")
    print("3 - ضرب")
    print("4 - قسمة")
    print("5 - خروج")

    choice = int(input("> "))

    if choice == 5:
        print("تم إنهاء البرنامج.")
        break

    num1 = float(input("أدخل الرقم الأول: "))
    num2 = float(input("أدخل الرقم الثاني: "))

    if choice == 1:
        print("النتيجة:", num1 + num2)
    elif choice == 2:
        print("النتيجة:", num1 - num2)
    elif choice == 3:
        print("النتيجة:", num1 * num2)
    elif choice == 4:
        if num2 != 0:
            print("النتيجة:", num1 / num2)
        else:
            print("خطأ: لا يمكن القسمة على صفر!")
    else:
        print("خيار غير صالح، حاول مجددًا.")
```

<div dir="rtl" style="text-align: right;">  

المهمة 3: متوسط – جمع الأرقام حتى يدخل المستخدم 0

</div>

```python
total = 0
num = int(input("أدخل رقم (0 للإنهاء): "))

while num != 0:
    total += num
    num = int(input("أدخل رقم (0 للإنهاء): "))

print("المجموع =", total)
```

المهمة 4: متقدم – التحقق من كلمة المرور مع 3 محاولات فقط

</div>

```python

password = "python123"
attempts = 0
max_attempts = 3
locked = True

while attempts < max_attempts:
    user_input = input(f"محاولة {attempts + 1}/{max_attempts} - أدخل كلمة المرور: ")
    attempts += 1
    
    if user_input == password:
        print("مرحبًا! تم تسجيل الدخول بنجاح.")
        locked = False
        break
if locked:
    print("فشلت جميع المحاولات. تم حظرك.")
```
