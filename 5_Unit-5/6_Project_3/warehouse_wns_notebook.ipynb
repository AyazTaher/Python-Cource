```python
warehouse = []

users = {
    "admin": {"password": "admin123", "role": "manager"},
    "viewer": {"password": "view123", "role": "viewer"}
}

# Login
username = input("Enter username: ")
password = input("Enter password: ")

if username not in users or users[username]["password"] != password:
    print("Access Denied! Incorrect username or password.")
    exit()
else:
    role = users[username]["role"]
    print(f"Access Granted! Logged in as {username.upper()} ({role.upper()}).")

# ===== MAIN MENU =====
while True:
    print("=== WAREHOUSE MENU ===")
    print("1.Add new item")
    print("2.View all items")
    print("3.Check total products and quantities")
    print("4.Receive items (increase stock)")
    print("5.Issue items (decrease stock)")
    print("6.Search for item by name")
    print("7.View low-stock items")
    print("8.Exit")

    choice = input("Choose an option (1-8): ")

    # Permission control (Viewer role cannot modify stock)
    if role == "viewer" and choice in ["1", "4", "5"]:
        print("Permission Denied!You do not have access to modify stock.")
        continue

    # 1.Add new item
    if choice == "1":
        name = input("Enter item name: ")
        desc = input("Enter item description: ")
        while True:
            quantity = int(input("Enter quantity: "))
            if quantity <= 0:
                print("Quantity must be greater than 0.Try again.")
            else:
                break
        item = {"name": name, "desc": desc, "quantity": quantity}
        warehouse.append(item)
        print(f"{name} added successfully!")

    # 2.View all items
    elif choice == "2":
        if len(warehouse) == 0:
            print("Warehouse is empty.")
        else:
            print("--- Current Inventory ---")
            for i, item in enumerate(warehouse, start=1):
                print(f"{i}.{item['name']} | {item['desc']} | Quantity: {item['quantity']}")

    # 3.General statistics
    elif choice == "3":
        if len(warehouse) == 0:
            print("Warehouse is empty.")
        else:
            total_items = len(warehouse)
            total_quantity = sum(item["quantity"] for item in warehouse)
            print(f"Total products: {total_items}")
            print(f"Total quantity: {total_quantity}")

    # 4.Increase stock
    elif choice == "4":
        if len(warehouse) == 0:
            print("No items in warehouse.")
        else:
            name = input("Enter item name to increase stock: ")
            found = False
            for item in warehouse:
                if item["name"].lower() == name.lower():
                    while True:
                        add_qty = int(input("Enter quantity to add: "))
                        if add_qty <= 0:
                            print("Quantity must be positive.")
                        else:
                            break
                    item["quantity"] += add_qty
                    print(f"Stock updated! {item['name']} now has {item['quantity']} units.")
                    found = True
                    break
            if not found:
                print("Item not found!")

    # 5.Decrease stock
    elif choice == "5":
        if len(warehouse) == 0:
            print("No items in warehouse.")
        else:
            name = input("Enter item name to issue: ")
            found = False
            for item in warehouse:
                if item["name"].lower() == name.lower():
                    while True:
                        issue_qty = int(input("Enter quantity to issue: "))
                        if issue_qty <= 0:
                            print("Quantity must be positive.")
                        else:
                            break
                    if issue_qty > item["quantity"]:
                        print("Not enough stock!")
                    else:
                        item["quantity"] -= issue_qty
                        print(f"Issued {issue_qty} units.Remaining: {item['quantity']}")
                    found = True
                    break
            if not found:
                print("Item not found!")

    # 6.Search items (partial name match)
    elif choice == "6":
        if len(warehouse) == 0:
            print("Warehouse is empty.")
        else:
            keyword = input("Enter part of item name to search: ").lower()
            found_any = False
            print("Search Results:")
            for item in warehouse:
                if keyword in item["name"].lower():
                    print(f"- {item['name']} | {item['desc']} | Qty: {item['quantity']}")
                    found_any = True
            if not found_any:
                print("No items match your search.")

    # 7.View low-stock items
    elif choice == "7":
        if len(warehouse) == 0:
            print("Warehouse is empty.")
        else:
            limit = int(input("Enter minimum stock limit (e.g., 5): "))
            if limit <= 0:
                print("Limit must be greater than 0.")
            else:
                found = False
                print(f"Items below {limit} units:")
                for item in warehouse:
                    if item["quantity"] < limit:
                        print(f"- {item['name']} | Qty: {item['quantity']}")
                        found = True
                if not found:
                    print("All items have sufficient stock.")

    # 8.Exit and export data
    elif choice == "8":
        if len(warehouse) > 0:
            with open("warehouse.csv", "w") as f:
                f.write("name,desc,quantity")
                for item in warehouse:
                    desc_safe = item["desc"].replace(",", ";")
                    f.write(f"\n{item['name']},{desc_safe},{item['quantity']}")
            print("Warehouse data exported to 'warehouse.csv'")
        print("Exiting warehouse system.Goodbye!")
        break

    else:
        print("Invalid choice.Please enter a number between 1 and 8.")

```
