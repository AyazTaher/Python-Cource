المهمة 1 – كلمة السر بثلاث محاولات

المطلوب:

1.	الكلمة الصحيحة هي "python123".
2.	اسمح للمستخدم بـ 3 محاولات فقط.
3.	إذا كانت صحيحة، اطبع "تم تسجيل الدخول بنجاح!".
4.	إذا أخطأ في الثلاث مرات، اطبع "تم حظر الحساب!".


</div>

```python

password = "python123"
tries = 0

while tries < 3:
    user_input = input("Enter the password: ")
    if user_input == password:
        print("Login successful!")
        break
    else:
        print("Incorrect password!Try again.")
        tries += 1

        if tries == 3:
            print("Account has been locked!")

            
```

المهمة 2 – آلة حاسبة تفاعلية
المطلوب:

1.	أنشئ قائمة خيارات:
2. جمع
3. طرح
4. ضرب
5. قسمة
6. خروج
-	اطلب من المستخدم اختيار رقم العملية.
-	أدخل رقمين، ثم نفّذ العملية المختارة.
-	استمر في الحلقة حتى يختار المستخدم رقم (5).



</div>

```python

choice = 0

while choice!= 5:
    print("" \
    "Choose an operation:")
    print("1 - Addition")
    print("2 - Subtraction")
    print("3 - Multiplication")
    print("4 - Division")
    print("5 - Exit")
    
    choice = int(input("> "))
    
    if choice == 5:
        print("Program terminated.")
        break
    
    num1 = float(input("Enter the first number: "))
    num2 = float(input("Enter the second number: "))
    
    if choice == 1:
        print("Result:", num1 + num2)
    elif choice == 2:
        print("Result:", num1 - num2)
    elif choice == 3:
        print("Result:", num1 * num2)
    elif choice == 4:
        if num2!= 0:
            print("Result:", num1 / num2)
        else:
            print("Error: Cannot divide by zero!")
    else:
        print("Invalid choice, try again.")
        
```

المهمة 3: متوسط – جمع الأرقام حتى يدخل المستخدم 0

</div>

```python

total = 0
num = int(input("Enter a number (0 to stop): "))

while num!= 0:
    total += num
    num = int(input("Enter a number (0 to stop): "))

    print("Total =", total)

    
```

المهمة 4: متقدم – التحقق من كلمة المرور مع 3 محاولات فقط

</div>

```python

password = "python123"
attempts = 0
max_attempts = 3
locked = True

while attempts < max_attempts:
    user_input = input(f"Attempt {attempts + 1}/{max_attempts} - Enter password: ")
    attempts += 1

    if user_input == password:
        print("Welcome!Login successful.")
        locked = False
        break

    if locked:
        print("All attempts failed.You have been locked out.")

        
```
