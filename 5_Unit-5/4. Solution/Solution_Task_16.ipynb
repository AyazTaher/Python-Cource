المهمة 3: متوسط إلى متقدم – طباعة جدول الضرب لعدد معين

اطلب من المستخدم إدخال رقم، ثم اطبع جدول الضرب له من 1 إلى 10


</div>

```python


num = int(input("Enter a number to display its multiplication table: "))

print(f"Multiplication table of {num}:")
for i in range(1, 11):
    print(f"{num}  {i} = {num * i}")


    
```

المهمة 4: متقدم – رسم مثلث من النجوم باستخدام حلقات متداخلة

 اطبع مثلثًا من النجوم كالتالي:
# *
# **
# ***
# ****
# *****

</div>

```python


for i in range(1, 6):  # Number of rows
    for j in range(i):  # Number of stars in each row
        print("*", end="")
        print("")  # New line

        
```
