التمرين 1: سهل – استخدام break مع while (إيقاف عند أول رقم سالب)

مثال إدخال: 5, 3, 8, -2 → يتوقف ويطبع 16
الحل:

</div>

```python

total = 0
while True:
    num = int(input("Enter a positive number (negative to stop): "))
    if num < 0:
        break  # Stop the loop immediately
    total += num
    print("Total =", total)

    

```

التمرين 2: سهل – استخدام continue مع for (تخطي الأرقام الزوجية)

المهمة: اطبع الأرقام من 1 إلى 10، لكن تخطَ الأرقام الزوجية.

الحل


</div>

```python

for i in range(1, 11):
    if i % 2 == 0:
        continue  # Skip the rest of this iteration
    print(i)

    
```
