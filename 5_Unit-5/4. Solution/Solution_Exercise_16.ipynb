التمرين 1: متوسط إلى متقدم – تحديد الأعداد الأولية (Prime Numbers) من 2 إلى 50

 اطبع جميع الأعداد الأولية من 2 إلى 50


 (عدد أولي: يقبل القسمة على 1 ونفسه فقط)

</div>

```python

for num in range(2, 51):
    is_prime = True
    for i in range(2, num):
        if num % i == 0:
            is_prime = False
            break
        if is_prime:
            print(num, end=" ")


            
```

التمرين 2: متقدم – تصنيف الطلاب حسب الدرجات (مع قائمة)
 لديك قائمة بالدرجات:
 
grades = [85, 92, 67, 45, 78, 90, 33, 88, 95, 60]

1.	صنّف الطلاب إلى:
2.	ممتاز: 90 أو أكثر
3.	جيد جدًا: 80–89
4.	جيد: 70–79
5.	مقبول: 60–69
6.	راسب: أقل من 60
7.	اطبع عدد الطلاب في كل فئة

</div>

```python


grades = [85, 92, 67, 45, 78, 90, 33, 88, 95, 60]
excellent = very_good = good = acceptable = fail = 0

for grade in grades:
    if grade >= 90:
        excellent += 1
    elif grade >= 80:
        very_good += 1
    elif grade >= 70:
        good += 1
    elif grade >= 60:
        acceptable += 1
    else:
        fail += 1

        print("Classification:")
        print(f"Excellent: {excellent}")
        print(f"Very Good: {very_good}")
        print(f"Good: {good}")
        print(f"Acceptable: {acceptable}")
        print(f"Fail: {fail}")

        
```
