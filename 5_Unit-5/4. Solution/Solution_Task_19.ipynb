المهمة 1 (متوسط)

المطلوب:
اكتب برنامجًا يقوم بتصفية القاموس ليطبع أسماء الطلاب الذين حصلوا على 80 أو أكثر مع درجاتهم.

البيانات المتاحة:

grades = {"Ali": 90, "Sara": 75, "Mona": 85, "Omar": 60}


</div>

```python

grades = {"Ali": 90, "Sara": 75, "Mona": 85, "Omar": 60}
for name, grade in grades.items():
    if grade >= 80:
        print(name, grade)

        

```

المهمة 2 (متوسط)

المطلوب:
اكتب برنامجًا يستخدم حلقة while لإزالة العناصر السالبة من قائمة ثم طباعة النتيجة النهائية.

البيانات المتاحة:

nums = [3, -1, 5, -4, 2, -6]


</div>

```python

# Removing negative numbers from a list

nums = [3, -1, 5, -4, 2, -6]
i = 0
while i < len(nums):
    if nums[i] < 0:
        nums.pop(i)
    else:
        i += 1
        print("After removal:", nums)

        
```

المهمة 3 (متقدم)

المطلوب:
اكتب برنامجًا لحساب تكرار كل قيمة داخل القائمة باستخدام قاموس، ثم طباعته.

البيانات المتاحة:

data = [1, 3, 1, 2, 3, 3, 4, 2, 1]


</div>

```python

data = [1, 3, 1, 2, 3, 3, 4, 2, 1]
count_dict = {}
for item in data:
    if item in count_dict:
        count_dict[item] += 1
    else:
        count_dict[item] = 1
        print(count_dict)
        
```

المهمة 4 (متقدم جدًا)

المطلوب:

اكتب برنامجًا يحدد ما إذا كان كل طالب ناجحًا أم راسبًا (النجاح إذا كانت الدرجة ≥ 60) ويطبع النتيجة بالصيغة:

students = [
    {"name": "Ali", "grade": 90},
    {"name": "Sara", "grade": 55},
    {"name": "Omar", "grade": 72},
    {"name": "Mona", "grade": 40},
]


</div>

```python

# Checking students' grades

students = [
    {"name": "Ali", "grade": 90},
    {"name": "Sara", "grade": 55},
    {"name": "Omar", "grade": 72},
    {"name": "Mona", "grade": 40},]

for student in students:
    if student["grade"] >= 60:
        print(f"{student['name']} passed with a grade of {student['grade']}")
    else:
        print(f"{student['name']} failed with a grade of {student['grade']}")
```
