المهمة 1: متوسط – else مع for (البحث عن رقم أولي)

المهمة: تحقق مما إذا كان الرقم 17 أوليًا باستخدام for + else.

- إذا وُجد مقسوم → ليس أوليًا
- إذا لم يُوجد → أولي

الحل 



</div>

```python

num = 17
is_prime = True

for i in range(2, num):
    if num % i == 0:
        print(f"{num} is not prime, it is divisible by {i}")
        is_prime = False
        break
else:
        print(f"{num} is a prime number!")

```

المهمة 2: متوسط – continue + while (تخطي الأرقام الفردية)

ابدأ من 1، اطبع الأرقام الزوجية فقط حتى 20، باستخدام while و continue.

الحل

</div>

```python

i = 1
while i <= 20:
    if i % 2!= 0:
        i += 1
        continue  # Skip odd numbers
    print(i)
    i += 1

    
```
