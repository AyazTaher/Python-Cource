المهمة 1 (متوسط): تحديد أكبر رقم

السؤال:

اكتب برنامجًا يطلب من المستخدم إدخال خمسة أرقام، ثم يستخدم حلقة for للبحث عن أكبر رقم بينها باستخدام شرط if.

استخدم while لجمع الإدخالات من المستخدم.


</div>

```python

# Finding the largest number among 5 user inputs

numbers = []
count = 0
while count < 5:
    num = int(input(f"Enter number {count + 1}: "))
    numbers.append(num)
    count += 1

    largest = numbers[0]
    for n in numbers:
        if n > largest:
            largest = n
            print("The largest number is:", largest)

            

```

المهمة 2 (متوسط): حساب عدد الأرقام الزوجية والفردية

السؤال:

اكتب برنامجًا يطلب من المستخدم إدخال 10 أرقام.

باستخدام حلقة while للإدخال وحلقة for داخلها، احسب عدد الأرقام الزوجية وعدد الأرقام الفردية.

في النهاية، اطبع النتيجة.الحل

</div>

```python

# Counting even and odd numbers from 10 user inputs

even = odd = 0
count = 0
while count < 10:
    num = int(input("Enter a number: "))
    for n in [num]:
        if n % 2 == 0:
            even += 1
        else:
            odd += 1
            count += 1

            print("Number of even numbers:", even)
            print("Number of odd numbers:", odd)

            
```
