التمرين 1 – العدّ من 1 إلى 10

المطلوب:

1.	اطبع عبارة "بدء العدّ...".
2.	استخدم حلقة while لعرض الأرقام من 1 إلى 10.
3.	بعد الانتهاء، اطبع "انتهى العدّ!".



</div>

```python

# Starting the count from 1 to 10 using a while loop

print("Starting the count...")

number = 1  # Starting from 1

# Loop continues as long as the number is less than or equal to 10
while number <= 10:
    print(number)

    # Increasing the number by 1 each time
    number += 1

    print("Counting finished!")

    
```

التمرين 2 – جمع الأرقام حتى إدخال صفر

 المطلوب:
 
1.	اطلب من المستخدم إدخال أرقام موجبة.
2.	استمر في الجمع حتى يُدخل المستخدم الرقم 0.
3.	عند إدخال صفر، اطبع مجموع الأرقام المدخلة.



</div>

<div dir="rtl" style="text-align: right;">  


</div>

```python

total = 0
num = int(input("Enter a number (0 to stop): "))

while num!= 0:
    total += num
    num = int(input("Enter a number (0 to stop):" ))

    print("The total sum is:", total)

    
```
