السيناريو العام 

أنت تعمل كمبرمج بيانات في مستشفى صغير.

لديك قائمة بيانات تحتوي على معلومات عن مجموعة من المرضى، تتضمن: العمر، الجنس، ضغط الدم، الكوليسترول، والدواء الذي يُستخدم لعلاج المريض.


البيانات </div>

```python
data = [
    {"Age": 23, "Sex": "F", "BP": "HIGH", "Cholesterol": "HIGH", "Drug": "drugY"},
    {"Age": 47, "Sex": "M", "BP": "LOW", "Cholesterol": "HIGH", "Drug": "drugC"},
    {"Age": 28, "Sex": "F", "BP": "NORMAL", "Cholesterol": "HIGH", "Drug": "drugX"},
    {"Age": 61, "Sex": "F", "BP": "LOW", "Cholesterol": "HIGH", "Drug": "drugY"}
]

print(data)
```

<div dir="rtl" style="text-align: right;">

<div dir="rtl" style="text-align: right;">

3-	while أو for مع range()</div>

<div dir="rtl" style="text-align: right;">
المطلوب: طباعة الأعمار باستخدام for مع range</div>

```python

for i in range(len(data)):
    print("Age:", data[i]["Age"])
    
```

<div dir="rtl" style="text-align: right;">
4-	استخدام while مع range:

```python

i = 0
while i < len(data):
    print("Age:", data[i]["Age"])
    i += 1
    
```

<div dir="rtl" style="text-align: right;">
5-	أوامر التحكم في الحلقات</div>

<div dir="rtl" style="text-align: right;">
- المطلوب: طباعة الدواء فقط للمرضى الذين ضغط دمهم HIGH، وتوقف بعد أول مريضين</div>

```python

count = 0
for patient in data:
    if patient["BP"]!= "HIGH":
        continue
    print("Age:", patient["Age"], "- Drug:", patient["Drug"])
    count += 1
    if count == 2:
        break
    
```

<div dir="rtl" style="text-align: right;">
6-	دمج if-elif-else مع while و for</div>

<div dir="rtl" style="text-align: right;">
- المطلوب: تصنيف المرضى حسب ضغط الدم</div>

```python

# Categorizing patients by blood pressure
for patient in data:
    if patient["BP"] == "HIGH":
        level = "High"
    elif patient["BP"] == "NORMAL":
        level = "Normal"
    else:
        level = "Low"
    print("Age:", patient["Age"], "- Blood Pressure:", level)
```

<div dir="rtl" style="text-align: right;">
7-	دمج الحلقات والشروط مع هياكل البيانات</div>

<div dir="rtl" style="text-align: right;">
- المطلوب: طباعة المرضى الذين ضغط دمهم HIGH ودواؤهم drugY</div>

```python
# Display patients with high BP taking drugY
for patient in data:
    if patient["BP"] == "HIGH" and patient["Drug"] == "drugY":
        print("Age:", patient["Age"], "- Sex:", patient["Sex"], "- Drug:", patient["Drug"])

```

<div dir="rtl" style="text-align: right;">

8-	تمرين حساب مجموع أعمار المرضى أقل من 50 سنة
</div>

```python

i = 0
sum_age = 0
while i < len(data):
    if data[i]["Age"] < 50:
        sum_age += data[i]["Age"]
    i += 1
    print("Sum of ages of patients under 50:", sum_age)
```

<div dir="rtl" style="text-align: right;">

9-	تمرين  طباعة الأعمار حتى نصل لمريض عمره أكبر من 60
</div>

```python

# Print ages until a patient older than 60 is found
i = 0
while i < len(data):
    if data[i]["Age"] > 60:
        break
    print("Age:", data[i]["Age"])
    i += 1
    
```

```python

# Count the number of patients taking each drug
drug_count = {}
for patient in data:
    if patient["Drug"] not in drug_count:
        drug_count[patient["Drug"]] = 1
    else:
        drug_count[patient["Drug"]] += 1

    print(drug_count)
```
