التمرين 1 – العدّ من 1 إلى 10

المطلوب:

1.	اطبع عبارة "بدء العدّ...".
2.	استخدم حلقة while لعرض الأرقام من 1 إلى 10.
3.	بعد الانتهاء، اطبع "انتهى العدّ!".



</div>

```python

# بدء العد من 1 إلى 10 باستخدام while loop

print("بدء العدّ...")

number = 1  # البداية من 1

# التكرار طالما الرقم أقل أو يساوي 10

while number <= 10:   
    print(number)
# زيادة الرقم بمقدار 1 في كل مرة

    number += 1      
print("انتهى العدّ!")
```

التمرين 2 – جمع الأرقام حتى إدخال صفر

 المطلوب:
 
1.	اطلب من المستخدم إدخال أرقام موجبة.
2.	استمر في الجمع حتى يُدخل المستخدم الرقم 0.
3.	عند إدخال صفر، اطبع مجموع الأرقام المدخلة.



</div>

<div dir="rtl" style="text-align: right;">  


</div>

```python

# جمع الأرقام حتى إدخال 0

# المتغير الذي سيحفظ المجموع
total = 0 
num = int(input("أدخل رقمًا (0 لإنهاء): "))

while num != 0:
    total += num
    num = int(input("أدخل رقمًا (0 لإنهاء): "))

print("مجموع الأرقام هو:", total)
```
