الدرس 5.5 – أوامر التحكم في الحلقات break, continue, else</div>

<div dir="rtl" style="text-align: right;">
أولًا: الأمر break</div>

<div dir="rtl" style="text-align: right;">
1-	استخدام break مع حلقة for</div>

```python

for num in range(1, 10):
    if num == 5:
        print("Number 5 found, ending the loop.")
        break
    print(num)
    
```

<div dir="rtl" style="text-align: right;">
2-	 استخدام break مع حلقة while</div>

```python

count = 0
while count < 10:
    if count == 4:
        print("Stopped at number 4")
        break
    print(count)
    count += 1
    
```

<div dir="rtl" style="text-align: right;">
ثانيًا: الأمر continue</div>

<div dir="rtl" style="text-align: right;">
1-	تخطي الأعداد الزوجية</div>

```python

for num in range(1, 6):
    if num % 2 == 0:
        continue
    print(num)
    
```

<div dir="rtl" style="text-align: right;">
2- استخدام  continue مع  while</div>

```python

x = 0
while x < 5:
    x += 1
    if x == 3:
        continue
    print(x)
    
```

<div dir="rtl" style="text-align: right;">
ثالثًا: else مع الحلقات</div>

<div dir="rtl" style="text-align: right;">
1-	 البحث عن عنصر</div>

```python

for num in range(1, 8):
    if num == 7:
        print("Number 7 found")
        break
    else:
        print("Number 7 not found")
```

<div dir="rtl" style="text-align: right;">
2-	 مع while</div>

```python

count = 0
while count < 3:
    print(count)
    count += 1
else:
    print("The loop finished normally")
    
```

<div dir="rtl" style="text-align: right;">
تمرين 
المطلوب:
اكتب برنامجًا يطلب من المستخدم إدخال أرقام صحيحة حتى يكتب الرقم 0

- إذا كان الرقم سالبًا، يتم تجاهله باستخدام continue
- إذا كان الرقم أكبر من 100، توقف البرنامج باستخدام break
- إذا انتهى البرنامج بشكل طبيعي (باستخدام 0)، اطبع رسالة "انتهى الإدخال بنجاح"

الحل </div>

```python

while True:
    num = int(input("Enter a number (0 to stop): "))

    if num < 0:
        print("Negative number ignored.")
        continue

    if num > 100:
        print("Program stopped because the number is greater than 100.")
        break

    if num == 0:
        print("Input finished successfully.")
        break
```
