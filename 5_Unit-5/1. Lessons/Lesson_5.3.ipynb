الدرس 5.3 – حلقة for </div>

<div dir="rtl" style="text-align: right;">
1-	 التكرار على قائمة</div>

```python

fruits = ["Apple", "Banana", "Orange"]
for fruit in fruits:
    print(f"I love eating {fruit}")
    
```

<div dir="rtl" style="text-align: right;">
2-	استخدام for مع range()</div>

```python

for i in range(1, 10):
    print("Number:", i)
```

<div dir="rtl" style="text-align: right;">
3-	التكرار على النصوص (Strings)</div>

```python

text = "Java Script"
for char in text:
    print(f" character: {char}")
```

<div dir="rtl" style="text-align: right;">
4-	 التكرار على قاموس (Dictionary)</div>

```python

person = {"name": "Ahmed", "age": 25, "city": "Damascus"}
for key, value in person.items():
    print(f"{key}: {value}")
    
```

<div dir="rtl" style="text-align: right;">
5-	 دمج for مع الشروط</div>

```python

numbers = [1, 2, 3, 4, 5]
for num in numbers:
    if num % 2 == 0:
        print(f"{num} even number")
    else:
        print(f"{num}  Odd number ")
        
```

<div dir="rtl" style="text-align: right;">
6- دمج for مع else</div>

```python

for i in range(6):
    print(i)
else:
    print("The loop ended without a break")
```

<div dir="rtl" style="text-align: right;">
7- دمج for مع enumerate</div>

```python

colors = ["Red", "Green", "Blue"]

for index, color in enumerate(colors):
    print(f"The color number {index} is {color}")
    
```

<div dir="rtl" style="text-align: right;">
تمرين 
المطلوب
اكتب برنامجًا يطلب من المستخدم إدخال 5 أسماء، ثم يطبعها مع أرقام ترتيبها (1 – 5)

الحل</div>

```python

names = []
for i in range(1, 6):
    name = input(f"Enter name number {i}: ")
    names.append(name)

    print("" \
    "List of names:")
    for index, name in enumerate(names, start=1):
        print(f"{index}.{name}")
```
