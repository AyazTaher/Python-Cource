<div dir="rtl" style="text-align: right;">
أولًا: الدمج بين while و if/else</div>

<div dir="rtl" style="text-align: right;">
1-	التحقق من الأرقام الموجبة والسالبة</div>

```python

# 1- Checking Positive and Negative Numbers
count = 0
while count < 5:
    num = int(input("Enter a number: "))

    if num >= 0:
        print("Positive number or zero")
    else:
        print("Negative number")

        count += 1
        
```

<div dir="rtl" style="text-align: right;">
2-	 while مع شرط إيقاف خاص</div>

```python

# 2- While loop with a custom stop condition
while True:
    word = input("Enter a word (type 'exit' to stop): ")

    if word == "exit":
        print("Program terminated.")
        break
    else:
        print(f"You typed: {word}")
        
```

<div dir="rtl" style="text-align: right;">
ثانيًا: الدمج بين for و if/else</div>

<div dir="rtl" style="text-align: right;">
1-	 تصنيف الأعداد إلى زوجية وفردية</div>

```python

# 1- Classifying numbers as even or odd
for num in range(1, 6):
    if num % 2 == 0:
        print(f"{num} is even")
    else:
        print(f"{num} is odd")
        
```

<div dir="rtl" style="text-align: right;">
2-	 فلترة قائمة أسماء</div>

```python

# 2- Filtering a list of names
names = ["Ali", "Sara", "Omar", "Huda"]

for name in names:
    if name.startswith("A"):
        print(f"{name} starts with the letter A")
    else:
        print(f"{name} does not start with the letter A")
        
```

<div dir="rtl" style="text-align: right;">
ثالثًا: الدمج بين while و for معًا + الشروط</div>

<div dir="rtl" style="text-align: right;">
1-	 البحث داخل قوائم متعددة</div>

```python

# 1- Searching inside multiple lists
lists = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

i = 0
while i < len(lists):
    for num in lists[i]:
        if num % 2 == 0:
            print(f"{num} is even")
        else:
            print(f"{num} is odd")
            i += 1
```

المطلوب
اكتب برنامجًا يطلب من المستخدم إدخال درجات طلاب (عددهم 3 طلاب). لكل طالب أدخل 4 مواد. البرنامج يجب أن

- يستخدم حلقة while للتكرار على الطلاب
- يستخدم حلقة for للتكرار على المواد.
- إذا كانت الدرجة أقل من 50، اطبع "راسب"
- إذا كانت الدرجة أكبر أو تساوي 50، اطبع "ناجح"

الحل </div>

```python

# Entering student grades and checking pass/fail
student_count = 0

while student_count < 3:
    print(f"Student number {student_count + 1}:")

    for subject in range(1, 5):
        grade = int(input(f"  Enter grade for subject {subject}: "))

        if grade >= 50:
            print("  Pass")
        else:
            print("  Fail")

            student_count += 1
            
```
