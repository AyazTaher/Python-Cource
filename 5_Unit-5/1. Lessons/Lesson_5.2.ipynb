<div dir="rtl" style="text-align: right;">
1-	 عدّاد بسيط</div>

```python

count = 1
while count <= 5:
    print("Loop number:", count)
    count += 1
    
```

<div dir="rtl" style="text-align: right;">
2-	 التحقق من إدخال كلمة مرور</div>

```python

password = ""
while password!= "python123":
    password = input("Enter the correct password: ")
    print("Login successful")
    
```

<div dir="rtl" style="text-align: right;">
3-	 جمع الأعداد من 1 حتى n</div>

```python

n = int(input("Enter a positive number: "))
total = 0
i = 1

while i <= n:
    total += i
    i += 1

    print(f"The sum of numbers from 1 to {n} is: {total}")
    
```

<div dir="rtl" style="text-align: right;">
4-	حلقة لانهائية مع كسرها بـ break</div>

```python

while True:
    name = input("Enter your name (type 'exit' to quit): ")
    if name.lower() == "exit":
        break
    print(f"Welcome, {name}!")
    
```

<div dir="rtl" style="text-align: right;">
5-	 استخدام continue لتجاوز دورة معينة</div>

```python

num = 1.5
while num < 10:
    num += 1
    if num % 2 == 0:
        continue
    print("Odd number:", num)
```

اكتب برنامجًا يطلب من المستخدم إدخال أرقام صحيحة، ويحسب المتوسط الحسابي لها، ويتوقف عن الإدخال عند كتابة الرقم -

```python

total = 0
count = 0

num = int(input("Enter a number (-1 to stop): "))
while num!= -1:
    total += num
    count += 1
    num = int(input("Enter a number (-1 to stop): "))

    if count > 0:
        average = total / count
        print(f"Number of entries: {count}")
        print(f"Average: {average}")
    else:
        print("No numbers were entered.")
        
```
