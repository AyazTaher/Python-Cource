<div dir="rtl" style="text-align: right;">
1-	 التكرار باستخدام for لطباعة الأرقام من 1 إلى 5</div>

```python

for i in range(1, 6):
    print(i)
    
```

<div dir="rtl" style="text-align: right;">
2-	 التكرار باستخدام while لحساب مجموع الأعداد</div>

```python

count = 1
total = 0

while count <= 5:
    total += count
    count += 1

    print("The Total is:", total)

    
```

<div dir="rtl" style="text-align: right;">
3-	 التكرار على قائمة أسماء</div>

```python

names = ["Ahmed", "Omer", "Sara"]

for name in names:
    print("Hello!", name)

    

```
