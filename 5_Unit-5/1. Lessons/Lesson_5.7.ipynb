أولًا: القوائم Lists</div>

<div dir="rtl" style="text-align: right;">
1-	 استخراج الأعداد الزوجية من قائمة</div>

```python

# 1- Extracting even numbers from a list
numbers = [3, 10, 15, 22, 7, 8]

for num in numbers:
    if num % 2 == 0:
        print(f"{num} is even")
    else:
        print(f"{num} is odd")
        
```

<div dir="rtl" style="text-align: right;">
تطوير الكود:
لنقم بتطوير السكربت بحيث يحفظ الاعداد الفردية ضمن قائمة و الاعداد الفردية ضمن قائمة
</div>

```python

# Extracting even and odd numbers into separate lists
numbers = [3, 10, 15, 22, 7, 8]

even_numbers = []  # List to store even numbers
odd_numbers = []   # List to store odd numbers

for num in numbers:
    if num % 2 == 0:
        even_numbers.append(num)  # Add even number to the list
        print(f"{num} is even")
    else:
        odd_numbers.append(num)   # Add odd number to the list
        print(f"{num} is odd")

        print("Even numbers:", even_numbers)
        print("Odd numbers:", odd_numbers)
```

<div dir="rtl" style="text-align: right;">
ثانيًا: الـ Tuples

<div dir="rtl" style="text-align: right;">
1-	 تصنيف درجات الطلاب</div>

```python

# 1- Classifying student grades
grades = (65, 48, 90, 77, 35)

for grade in grades:
    if grade >= 50:
        print(f"{grade} Pass")
    else:
        print(f"{grade} Fail")
        
```

<div dir="rtl" style="text-align: right;">
تطوير الكود
لنقم بتطوير السكربت بحيث يحفظ الدرجات الناجحة ضمن قائمة و الدرجات الراسبة ضمن قائمة</div>


```python

# Classifying student grades into passed and failed lists
grades = (65, 48, 90, 77, 35)

passed = []   # List to store passed grades
failed = []   # List to store failed grades

for grade in grades:
    if grade >= 50:
        passed.append(grade)   # Add passed grade to the list
        print(f"{grade} Pass")
    else:
        failed.append(grade)   # Add failed grade to the list
        print(f"{grade} Fail")

        print("Passed grades:", passed)
        print("Failed grades:", failed)
```

<div dir="rtl" style="text-align: right;">
ثالثًا: الـ Sets</div>

<div dir="rtl" style="text-align: right;">
1-	 التحقق من وجود قيم محددة</div>

```python

# 1- Checking for specific values in a set
unique_numbers = {2, 4, 6, 8, 10}
check_list = [3, 4, 7, 8]

for num in check_list:
    if num in unique_numbers:
        print(f"{num} exists in the set")
    else:
        print(f"{num} does not exist")
        
```

<div dir="rtl" style="text-align: right;">
تطوير الكود:
لنقم بتطوير السكربت بحيث تخزين الأعداد الموجودة في المجموعة  ضمن قائمة و لتخزين الأعداد غير الموجودة
</div>

```python

# Checking numbers and storing found and not found numbers
unique_numbers = {2, 4, 6, 8, 10}
check_list = [3, 4, 7, 8]

found = []      # List to store numbers found in the set
not_found = []  # List to store numbers not found in the set

for num in check_list:
    if num in unique_numbers:
        found.append(num)      # Add found number to the list
        print(f"{num} exists in the set")
    else:
        not_found.append(num)  # Add not found number to the list
        print(f"{num} does not exist")

        print("Numbers found:", found)
        print("Numbers not found:", not_found)
```

<div dir="rtl" style="text-align: right;">
رابعًا: القواميس Dictionaries</div>

<div dir="rtl" style="text-align: right;">
1-	 تحديد المنتجات باهظة الثمن</div>

```python

# 1- Identifying expensive products
products = {
    "Laptop": 1500,
    "Phone": 800,
    "Headphones": 50
}

for item, price in products.items():
    if price > 1000:
        print(f"{item} is expensive")
    else:
        print(f"{item} has a reasonable price")
```

<div dir="rtl" style="text-align: right;">
تطوير الكود
لنقم بتطوير السكربت بحيث يتم تخزين المنتجات الباهظة الثمن ضمن قائمة و تخزين المنتجات ذات السعر المناسب ضمن قائمة </div>

```python

# Classifying products by price
products = {
    "Laptop": 1500,
    "Phone": 800,
    "Headphones": 50
}

expensive = []   # List to store expensive products
affordable = []  # List to store reasonably priced products

for item, price in products.items():
    if price > 1000:
        expensive.append(item)   # Add expensive product to the list
        print(f"{item} is expensive")
    else:
        affordable.append(item)  # Add reasonably priced product to the list
        print(f"{item} has a reasonable price")

        print("Expensive products:", expensive)
        print("Reasonably priced products:", affordable)
```

```python

# Classifying students by grade and checking age
students = [
    {"name": "Ali", "age": 17, "grade": 80},
    {"name": "Sara", "age": 19, "grade": 45},
    {"name": "Omar", "age": 16, "grade": 55}
]
print(students)

for student in students:
    if student["grade"] >= 50:
        print(f"{student['name']} Pass")
    else:
        print(f"{student['name']} Fail")

    if student["age"] < 18:
        print(f"{student['name']} is under 18 years old")

```
