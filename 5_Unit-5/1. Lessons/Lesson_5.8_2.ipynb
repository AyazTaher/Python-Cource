<div dir="rtl" style="text-align: right;">
-	إنشاء القائمة</div>

```python

cities = ["London", "Paris", "Rome", "Berlin", "Madrid"]

print(cities)
```

<div dir="rtl" style="text-align: right;">
-	تعديل اسم المدينة الثالثة</div>

```python

cities[2] = "Athens"
print(cities)
```

<div dir="rtl" style="text-align: right;">
-	إضافة مدينتين</div>

```python

cities.extend(["Lisbon", "Oslo"])

print(cities)
```

<div dir="rtl" style="text-align: right;">
-	الطباعة باستخدام for</div>

```python

print("Cities using for loop:")
for city in cities:
    print(city)

    
```

<div dir="rtl" style="text-align: right;">
-	الطباعة باستخدام enumerate</div>

```python

print("Cities using enumerate:")
for index, city in enumerate(cities, start=1):
    print(f"City {index}: {city}")

    
```
