مثال– إدارة قائمة طلاب صف دراسي</div>

<div dir="rtl" style="text-align: right;">
السيناريو</div>

<div dir="rtl" style="text-align: right;">
لدينا قائمة بأسماء طلاب صف، ونريد أن:

1.	ننشئ القائمة ونضيف إليها أنواع بيانات مختلفة (أسماء، أعمار).
2.	نعرض عناصر محددة باستخدام الفهرسة والتقطيع.
3.	نعدل بعض القيم داخل القائمة.
4.	نستخدم دوال القوائم المدمجة للإضافة والحذف والتوسيع.
5.	نكرر على القائمة لعرض بيانات الطلاب بطرق مختلفة.</div>


<div dir="rtl" style="text-align: right;">
الحل</div>

<div dir="rtl" style="text-align: right;">
1- إنشاء القوائم وأنواع العناصر

- students تحتوي على أسماء الطلاب
- ages تحتوي على أعمارهم
- mixed_list مثال لقائمة تحتوي أنواع بيانات مختلفة
</div>

```python

# 1- Creating lists and different element types

# List of names
students = ["Ali", "Sara", "Omar"]

# List of ages
ages = [20, 22, 19]

# Mixed-type list
mixed_list = ["Lama", 21, True, 85.5]

print("Students list:", students)
print("Ages list:", ages)
print("Mixed list:", mixed_list)
```

<div dir="rtl" style="text-align: right;">
2- الفهرسة والتقطيع

- [0] للوصول لأول عنصر.
- [-1] للوصول لآخر عنصر.
- [0:2] لعرض أول عنصرين.
</div>

```python

# Indexing
print("First student:", students[0])
print("Last student:", students[-1])

# Slicing
print("First two students:", students[0:2])
```

<div dir="rtl" style="text-align: right;">
3-	تعديل القوائم

- نستطيع تغيير عنصر واحد مثل students[1] = "Mona".
- أو تعديل مجموعة عناصر دفعة واحدة عبر التقطيع.
</div>

```python

# Modify the second student's name
students[1] = "Mona"
print("After modifying the second name:", students)

# Modify more than one element using slicing
students[0:2] = ["Yousef", "Rami"]
print("After modifying the first two students:", students)
```

<div dir="rtl" style="text-align: right;">
4-	دوال القوائم المدمجة

- append() لإضافة عنصر في النهاية.
- remove() لحذف عنصر بالقيمة.
- pop() لحذف عنصر حسب الفهرس وإرجاعه.
- extend() لدمج قائمة بأخرى.
- insert() لإضافة عنصر في موقع محدد.
- clear() لتفريغ القائمة (اختياري).
</div>

```python

# Add an element to the end
students.append("Laila")
print("After append:", students)

# Remove an element by value
students.remove("Rami")
print("After remove:", students)

# Remove by index and return it
removed_student = students.pop(1)
print(f"After pop, removed: {removed_student}")
print("Current list:", students)

# Extend the list with another list
students.extend(["Hana", "Khaled"])
print("After extend:", students)

# Insert an element at a specific index
students.insert(1, "Nour")
print("After insert:", students)

# Clear the list (optional)
# students.clear()
print("---")

```

<div dir="rtl" style="text-align: right;">
5-	التكرار على القوائم

- for: أبسط وأوضح طريقة للمرور على العناصر
- while: تعطيك تحكمًا أكبر بالفهرس
- enumerate(): تجمع الفهرس والقيمة معًا
</div>

```python

# Using for loop
print("Iteration using for:")
for student in students:
    print(f"Student: {student}")

# Using while loop
print("Iteration using while:")
i = 0
while i < len(students):
    print(f"Student {i+1}: {students[i]}")
    i += 1


# Using enumerate to show index and name together
print("Iteration using enumerate:")
for index, student in enumerate(students, start=1):
    print(f"Student number {index}: {student}")
```
