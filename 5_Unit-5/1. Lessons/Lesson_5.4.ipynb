الدرس 5.4 – الدالة range() واستخدامها في بايثون</div>

<div dir="rtl" style="text-align: right;">
1-	 استخدام range(stop)</div>

```python

for i in range(5):
    print(i)
    
```

<div dir="rtl" style="text-align: right;">
2-	 استخدام range(start, stop)</div>

```python

for i in range(2, 7):
    print(i)
    
```

<div dir="rtl" style="text-align: right;">
3-	 استخدام range(start, stop, step)</div>

```python

for i in range(0, 11, 2):
    print(i)
    
```

<div dir="rtl" style="text-align: right;">
4-	: العد التنازلي باستخدام step سالب</div>

```python

for i in range(10, 0, -2):
    print(i)
    
```

<div dir="rtl" style="text-align: right;">
5-	 تحويل النتيجة إلى قائمة</div>

```python

nums = list(range(1, 6))
print(nums)
```
