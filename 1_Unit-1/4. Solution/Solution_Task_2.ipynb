<div dir="rtl" style="text-align: right;">
المهمة 1: تحليل نص طويل
المطلوب:

1.	حوّل النص إلى حروف صغيرة.
2.	احسب كم مرة ظهرت كلمة "easy".
3.	استبدل كلمة "easy" بـ "simple".
4.	احسب عدد الحروف الكلي (بدون فراغات).
5.	قسم النص إلى قائمة كلمات.
6.	اطبع أول وآخر كلمة من النص.
7.	أعد تركيب الجملة بحيث تكون الكلمات مفصولة بـ _ (شرطة سفلية).
8.	اطبع طول الجملة الجديدة.

  </div>

```python

# Long text (you can replace it with any text)
text = "Python is an easy programming language.Learning Python is easy and fun.It makes solving problems easy and straightforward."

# 1.Convert text to lowercase
text_lower = text.lower()

# 2.Count how many times the word "easy" appears
count_easy = text_lower.count("easy")

# 3.Replace the word "easy" with "simple"
text_replaced = text_lower.replace("easy", "simple")

# 4.Count total characters (excluding spaces)
text_no_spaces = text_replaced.replace(" ", "")
total_chars = len(text_no_spaces)

# 5.Split the text into a list of words
words_list = text_replaced.split()

# 6.Print the first and last word of the text
first_word = words_list[0]
last_word = words_list[-1]

# 7.Reconstruct the sentence with words separated by "_"
new_sentence = "_".join(words_list)

# 8.Print the length of the new sentence
new_sentence_length = len(new_sentence)

# Final output
print("Text in lowercase:\n", text_lower)
print("Number of times 'easy' appears:", count_easy)
print("Text after replacing 'easy' with 'simple':\n", text_replaced)
print("Total characters (excluding spaces):", total_chars)
print("List of words:\n", words_list)
print("First word:", first_word)
print("Last word:", last_word)
print("New sentence with words separated by '_':\n", new_sentence)
print("Length of the new sentence:", new_sentence_length)
```

	   المهمة 2: تنظيف بيانات نصية خام (متقدم)

المطلوب:
1.	احذف الفراغات من البداية والنهاية.
2.	حوّل النص إلى حروف صغيرة.
3.	اجعل الكلمة الأولى فقط تبدأ بحرف كبير.
4.	احسب عدد الأحرف بعد التنظيف.
5.	تحقق إن كان النص يحتوي على كلمة "best".
6.	استبدل كلمة "best" بكلمة "most powerful".
7.	قسم النص إلى كلمات واطبعها كقائمة.
8.	أعد تركيب النص من جديد بفاصل فراغ واحد فقط.
9.	اطبع النص النهائي بشكل منسق كالتالي:

</div>

```python

# Raw text (you can replace it with your own text)
text = "   *Python 3 is the BEST programming-language!"

# 1.Remove spaces from the beginning and end
text = text.strip()

# 2.Convert text to lowercase
text = text.lower()

# 3.Capitalize only the first word
text = text.capitalize()

# 4.Count the number of characters after cleaning
num_chars = len(text.replace(" ", ""))  # exclude spaces

# 5.Check if the text contains the word "best"
contains_best = "best" in text.lower()

# 6.Replace "best" with "most powerful"
text = text.replace("best", "most powerful")

# 7.Split the text into words and print as a list
words_list = text.split()

# 8.Reconstruct the text with a single space between words
clean_text = " ".join(words_list)

# 9.Printt he final text in a formatted way
print("Final cleaned text:", clean_text)
print("Number of characters (without spaces):", num_chars)
print("Contains the word 'best'?:", contains_best)
print("Words list:", words_list)


```
