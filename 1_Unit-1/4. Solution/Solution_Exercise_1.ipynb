التمرين 1: العمليات الحسابية الأساسية
المطلوب:

1.	أنشئ متغيرين يحتويان على رقمين من اختيارك (مثلًا a = 8 و b = 3).
2.	احسب ناتج الجمع، الطرح، القسمة، الضرب، والأس البسيط بينهما.
3.	تحقق من أن الرقم الأول أكبر من الثاني.
4.	تحقق إذا كان كلا الرقمين فرديين.
5.	اطبع جميع النتائج مع توضيحها.
 الحل:
</div>

<div dir="rtl" style="text-align: right;">
الحل</div>

```python

# Create variables
a = 8
b = 3

# Arithmetic operations
sum_result = a + b
diff_result = a - b
prod_result = a * b
div_result = a / b
power_result = a ** b

# Conditional check
is_a_greater = a > b

# Output
print("Sum:", sum_result)
print("Difference:", diff_result)
print("Product:", prod_result)
print("Division:", div_result)
print("Power:", power_result)
print("Is the first greater?", is_a_greater)

```

<div dir="rtl" style="text-align: right;">  التمرين 2: الجمع بين الأعداد والشروط المتداخلة
المطلوب:

1.	أنشئ ثلاثة متغيرات رقمية (x, y, z).
2.	احسب متوسط الأرقام الثلاثة.
3.	تحقق إذا كان المتوسط أكبر من 10.
4.	إذا كان المتوسط أكبر من 10، اطبع "القيمة مرتفعة"،
وإذا كان أقل من 5 اطبع "القيمة منخفضة"،
غير ذلك اطبع "القيمة متوسطة".
5.	تحقق أيضًا مما إذا كانت جميع الأرقام موجبة.</div>

```python

x = 5
y = 12
z = 8

# Calculate the average
average = (x + y + z) / 3

# Print the average
print("Average =", average)

# Nested conditions
if average > 10:
    print("The value is high")
elif average < 5:
    print("The value is low")
else:
    print("The value is moderate")

# Check if all numbers are positive
all_positive = (x > 0) and (y > 0) and (z > 0)
print("Are all numbers positive?", all_positive)

```
