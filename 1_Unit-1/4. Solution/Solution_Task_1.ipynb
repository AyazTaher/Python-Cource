المهمة 1: العمل مع القيم المنطقية والمقارنة المركّبة
المطلوب:

1.	أنشئ متغيرين: num1 = 15 و num2 = 4.
2.	تحقق من الشروط التالية:

- هل num1 قابل للقسمة على num2 بدون باقي؟
- 	هل num1 أكبر من 10 وnum2 أصغر من 10؟
- 	هل أحد الرقمين فقط زوجي (باستخدام XOR المنطقي !=)؟
3.	اطبع النتائج بوضوح.
  </div>

```python

num1 = 15
num2 = 4

# Check divisibility
divisible = num1 % num2 == 0

# Double condition
condition = (num1 > 10) and (num2 < 10)

# Check if only one of them is even
one_even = (num1 % 2 == 0)!= (num2 % 2 == 0)

# Print results
print("Is", num1, "divisible by", num2, "?", divisible)
print("Is", num1, "> 10 and", num2, "< 10?", condition)
print("Is only one of the numbers even?", one_even)
```
