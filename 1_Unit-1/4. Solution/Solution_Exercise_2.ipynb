<div dir="rtl" style="text-align: right;">  التمرين 2: تنظيف نص يحتوي على رموز وأرقام
المطلوب:

1.	احذف الرموز (# و !).
3.	حوّل النص إلى أحرف صغيرة.
4.	استبدل كلمة "amazing" بكلمة "useful".
5.	تحقق إن كان النص يحتوي على كلمة "data".
6.	احسب عدد الكلمات في النص بعد التنظيف.
7.	اطبع النص بعد التعديلات.
</div>

```python
text = "### Data123 Science 456 is AMAZING!!! ###"
# 1. Remove '#' and '!'
text = text.replace("#", "").replace("!", "")
# 2. Convert text to lowercase
text = text.lower()
# 3. Replace the word "amazing" with "useful"
text = text.replace("amazing", "useful")
# 4. Check if the text contains the word "data"
has_data = "data" in text
# 5. Count the number of words after cleaning
words = text.split()
count = len(words)
# 6. Print results
print("Text after cleaning:", text)
print("Number of words:", count)
print("Does it contain the word 'data'?", has_data)
```
