<div dir="rtl" style="text-align: right;">
المهمة 1: تحليل نص طويل
المطلوب:

1.	حوّل النص إلى حروف صغيرة.
2.	احسب كم مرة ظهرت كلمة "easy".
3.	استبدل كلمة "easy" بـ "simple".
4.	احسب عدد الحروف الكلي (بدون فراغات).
5.	قسم النص إلى قائمة كلمات.
6.	اطبع أول وآخر كلمة من النص.
7.	أعد تركيب الجملة بحيث تكون الكلمات مفصولة بـ _ (شرطة سفلية).
8.	اطبع طول الجملة الجديدة.

  </div>

```python

paragraph = "Python is easy to learn, easy to use, and easy to read."

# 1. تحويل النص إلى حروف صغيرة
paragraph = paragraph.lower()

# 2. حساب عدد مرات ظهور "easy"
count_easy = paragraph.count("easy")

# 3. استبدال كلمة "easy" بـ "simple"
paragraph = paragraph.replace("easy", "simple")

# 4. حساب عدد الحروف بدون فراغات
no_spaces = len(paragraph.replace(" ", ""))

# 5. تقسيم النص إلى قائمة كلمات
words = paragraph.replace(",", "").replace(".", "").split()

# 6. طباعة أول وآخر كلمة
first_word = words[0]
last_word = words[-1]

# 7. إعادة تركيب الجملة باستخدام "_"
new_sentence = "_".join(words)

# 8. طباعة الطول النهائي
print("النص الجديد:", new_sentence)
print("عدد مرات easy:", count_easy)
print("عدد الحروف بدون فراغات:", no_spaces)
print("أول كلمة:", first_word)
print("آخر كلمة:", last_word)
print("طول النص النهائي:", len(new_sentence))

```

<div dir="rtl" style="text-align: right;">
	   المهمة 2: تنظيف بيانات نصية خام (متقدم)

المطلوب:
1.	احذف الفراغات من البداية والنهاية.
2.	احذف جميع الأرقام والرموز الخاصة (* و - و !).
3.	حوّل النص إلى حروف صغيرة.
4.	اجعل الكلمة الأولى فقط تبدأ بحرف كبير.
5.	احسب عدد الأحرف بعد التنظيف.
6.	تحقق إن كان النص يحتوي على كلمة "best".
7.	استبدل كلمة "best" بكلمة "most powerful".
8.	قسم النص إلى كلمات واطبعها كقائمة.
9.	أعد تركيب النص من جديد بفاصل فراغ واحد فقط.
10.	اطبع النص النهائي بشكل منسق كالتالي:

</div>

```python

raw_text = "   123***Python---is---THE---Best!!!***456   "

# 1. حذف الفراغات
clean = raw_text.strip()

# 2. حذف الأرقام
for digit in "0123456789":
    clean = clean.replace(digit, "")

# 3. حذف الرموز الخاصة
for symbol in ["*", "-", "!"]:
    clean = clean.replace(symbol, "")

# 4. تحويل النص إلى أحرف صغيرة
clean = clean.lower()

# 5. جعل الكلمة الأولى فقط تبدأ بحرف كبير
clean = clean.capitalize()

# 6. حساب عدد الأحرف بعد التنظيف
char_count = len(clean)

# 7. التحقق إن كان يحتوي على كلمة "best"
contains_best = "best" in clean

# 8. استبدال "best" بـ "most powerful"
clean = clean.replace("best", "most powerful")

# 9. تقسيم النص إلى كلمات
words = clean.split()

# 10. إعادة تركيب النص بفاصل فراغ
final_text = " ".join(words)

# 11. طباعة النص بشكل منسق
print(f"Clean text: {final_text}")
print(f"Word count: {len(words)}")
print(f"Contains 'best': {contains_best}")
print(f"Character count: {char_count}")
```
