<div dir="rtl" style="text-align: right;">
التمرين التطبيقي
لديك النص التالي:

sentence = "   python programming is FUN and powerful 123   "

المطلوب

1.	احذف الفراغات من البداية والنهاية
2.	اجعل النص كله بحروف صغيرة
3.	اجعل أول كلمة يبدأ حرفها كبيرًا فقط
4.	استبدل كلمة "FUN" بكلمة "interesting"
5.	احسب كم مرة تكررت كلمة "python"
6.	تحقق هل النص يبدأ بكلمة "python"
7.	تحقق هل النص ينتهي بأرقام
8.	قسم النص إلى قائمة كلمات
9.	اجمع القائمة السابقة من جديد في جملة مفصولة بشرطة 
10.	اطبع الحرف الأول والحرف الأخير من النص بعد التنظيف

الحل
</div>

```python
sentence = "   python programming is FUN and powerful 123   "

# 1- إزالة الفراغات
cleaned = sentence.strip()
print("After strip", cleaned)

# 2- تحويل لحروف صغيرة
lowered = cleaned.lower()
print("حروف صغيرة",lowered )

# 3- جعل أول كلمة بحرف كبير
capitalized = lowered.capitalize()
print("اول حرف كبير", capitalized)

# 4- استبدال كلمة
replaced = cleaned.replace("FUN", "interesting")
print("بعد الاستبدال: ", replaced)

# 5- عدد مرات تكرار كلمة python
print("عدد مرات تكررار python:",lowered.count("python") )

# 6- التحقق من البداية
print(" هل يبدا ب python?", lowered.startswith("python"))

# 7- التحقق من النهاية
print("ينتهي بارقام؟", cleaned.endswith("123"))

# 8- تقسيم النص
words = cleaned.split()
print("قائمة بالكلمات", words)

# 9- دمج القائمة بشرطة
joined = "-".join(words)
print("دمج الكلمات بمسافة", joined)

# 10- الحرف الأول والأخير
print("الحرف الاول", cleaned[0])
print("الحرف الأخير", cleaned[-1])
```
