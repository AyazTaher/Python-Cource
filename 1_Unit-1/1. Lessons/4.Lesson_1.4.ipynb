2-	تغيير قيمة المتغير أثناء التنفيذ
</div>

```python
x = 10

print("Initial value:", x)

x = x + 5

print("Updated value:", x)


```

<div dir="rtl" style="text-align: right;">
3-type()التعرف على نوع المتغير باستخدام 
</div>

```python
a = 42
b = 3.14
c = "Python"

print(type(a))
print(type(b))
print(type(c))


```

<div dir="rtl" style="text-align: right;">
4-	التحويل بين الأنواع
</div>

<div dir="rtl" style="text-align: right;">
-	تحويل عدد صحيح إلى نص
</div>

```python
num = 100
num_str = str(num)

print(num_str, type(num_str))


```

<div dir="rtl" style="text-align: right;">
-	تحويل نص إلى عدد صحيح
</div>

```python
age_str = "25"
age_int = int(age_str)

print(age_int, type(age_int))

```

<div dir="rtl" style="text-align: right;">

```python
print(bool(0))
print(bool(5))
print(bool(""))
print(bool("Hi"))


```
