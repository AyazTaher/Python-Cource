```python
num1 = input(" ادخل العدد الاول")
num2 = input(" ادخل الرقم الثاني")

# تحويل النصوص الى اعداد صحيحة
num1 = int(num1)
num2 = int(num2)

result = num1 + num2

print("مجموع العددين هو:", result)



```
