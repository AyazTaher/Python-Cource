1. العمليات الحسابية

```python
a = 10
b = 3

print("Addition:", a + b)
print("Substraction:", a - b)
print("Multiolicationl", a * b)
print("Division:", a / b)
print("Floor Division:", a // b)
print("Power:", a ** b)


```

<div dir="rtl" style="text-align: right;">
2. عمليات المقارنة
</div>

```python
x = 7
y = 5

print("Equal:", x==y)
print("Not Equal:", x!= y)
print("Greater:", x > y)
print("Less:", x < y)
print("Greater or Equal:", x >= y)
print("Less or Equal:", x <= y)

```

```python
age = 20
has_id = True

print("Can Enter:", (age >= 18) and has_id)

print("Discount:", (age < 18 ) or ( age > 60))

print("Not Adult:", ( age >= 18))





```
