3-	الطباعة باستخدام format()

```python
name = "laila"
age = 25
score = 95.75

print(" The Name: {}, The Age{}, The Score{}".format(name,age,score))


```

<div dir="rtl" style="text-align: right;">
4-	الطباعة باستخدام % formatting

```python
name = "laila"
age = 25
score = 95.75


print("Name:%s, Age:%d, Score:%.1f" % (name, age, score))



```

تمرين 

اكتب برنامجًا يحتوي على البيانات التالية:

- الاسم: "Ahmed"
- العمر: 28
- الراتب الشهري: 2500.75

المطلوب:
1.	أضف تعليقات توضح ما يفعله كل جزء من البرنامج.
2.	اطبع البيانات باستخدام الأساليب الثلاثة للطباعة:
- f-strings
- format()
- % formatting

الحل
</div>

```python
# تعريف المتغيرات
name = "Ahmed"
age = 28
monthly_salary = 2500.75


# الطباعة باستخدام f-strıng

print(f"Name: {name}, Age: {age}, Score: {monthly_salary}")


# الطباعة باستخدام  format()

print("Name: {}, Age: {}, Salary: {}".format(name, age, monthly_salary))



# الطباغة باستخدام  formatting %

print(" Name: %s, Age: %d, Salary: %.2f" % ( name, age, monthly_salary))






```
