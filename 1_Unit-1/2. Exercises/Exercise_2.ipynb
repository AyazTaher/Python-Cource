<div dir="rtl" style="text-align: right;">
التمرين 1: تنظيف جملة بسيطة

المطلوب:
1.	احذف الفراغات من البداية والنهاية.
2.	حوّل النص إلى أحرف صغيرة.
3.	احذف جميع علامات التعجب (!) والفواصل (,).
4.	اجعل أول حرف في الجملة كبيرًا فقط.
5.	تحقق إن كانت الكلمة "python" موجودة داخل النص.
6.	عدّ عدد الأحرف الموجودة في النص بعد التنظيف.
7.	اطبع النص النهائي.

</div>

<div dir="rtl" style="text-align: right;">
الحل</div>

```python

sentence = "   Hello, PYTHON world!!!   "

# 1. حذف الفراغات من البداية والنهاية
sentence = sentence.strip()

# 2. تحويل النص إلى أحرف صغيرة
sentence = sentence.lower()

# 3. حذف علامات التعجب (!) والفواصل (,)
sentence = sentence.replace("!", "").replace(",", "")

# 4. جعل أول حرف كبير فقط
sentence = sentence.capitalize()

# 5. التحقق إن كانت الكلمة "python" موجودة
contains_python = "python" in sentence

# 6. حساب عدد الأحرف بعد التنظيف
length = len(sentence)

# 7. طباعة النتائج
print("النص النهائي:", sentence)
print("هل يحتوي على كلمة python؟", contains_python)
print("عدد الأحرف:", length)

```

<div dir="rtl" style="text-align: right;">  التمرين 2: تنظيف نص يحتوي على رموز وأرقام
المطلوب:

1.	احذف الرموز (# و !).
2.	احذف الأرقام من النص تمامًا.
3.	حوّل النص إلى أحرف صغيرة.
4.	استبدل كلمة "amazing" بكلمة "useful".
5.	تحقق إن كان النص يحتوي على كلمة "data".
6.	احسب عدد الكلمات في النص بعد التنظيف.
7.	اطبع النص بعد التعديلات.
</div>

```python

text = "### Data123 Science 456 is AMAZING!!! ###"

# 1. حذف الرموز #
text = text.replace("#", "").replace("!", "")

# 2. حذف الأرقام من النص
for digit in "0123456789":
    text = text.replace(digit, "")

# 3. تحويل النص إلى أحرف صغيرة
text = text.lower()

# 4. استبدال كلمة "amazing" بـ "useful"
text = text.replace("amazing", "useful")

# 5. التحقق إن كان النص يحتوي على كلمة "data"
has_data = "data" in text

# 6. حساب عدد الكلمات بعد التنظيف
words = text.split()
count = len(words)

# 7. طباعة النتائج
print("النص بعد التنظيف:", text)
print("عدد الكلمات:", count)
print("هل يحتوي على كلمة 'data'؟", has_data)

```
