3.	احسب متوسط الدرجات لكل طالب وأضف المتوسط كمفتاح جديد باسم "average".

```python
student1 = {"name": "Ali", "age": 20, "grades": [85, 90, 78]}
student2 = {"name": "Sara", "age": 22, "grades": [92, 88, 95]}
student3 = {"name": "Omar", "age": 21, "grades": [70, 75, 80]}

students = [student1, student2, student3]

subjects = ("Math", "Physics", "Chemistry")
languages = {"Python", "Java", "C++"}

# 1
print(student1["name"])
print(student2["name"])
print(student3["name"])

# 2
student1["grades"].append(100)
student2["grades"].append(100)
student3["grades"].append(100)

# 3
student1["average"] = sum(student1["grades"]) / len(student1["grades"])
student2["average"] = sum(student2["grades"]) / len(student2["grades"])
student3["average"] = sum(student3["grades"]) / len(student3["grades"])

# 4
names_tuple = (student1["name"], student2["name"], student3["name"])
print(names_tuple)

# 5
languages.add("JavaScript")

# 6
languages.remove("C++")

# 7
print(students)
print(languages)
```

المطلوب:

2.	أضف مفتاح "genre" لكل كتاب: "Programming" للكتاب الأول والثالث، "Data Science" للكتاب الثاني.
3.	زد عدد الصفحات في الكتاب الثالث إلى 320.
4.	احذف مفتاح "author" من الكتاب الثاني.
5.	أضف "Machine Learning" إلى مجموعة genres.
6.	احذف "Data Science" من مجموعة genres.
7.	اطبع المكتبة ومجموعة الأنواع بعد التعديلات.

```python
book1 = {"title": "Python Basics", "author": "Ali", "pages": 350}
book2 = {"title": "Data Science", "author": "Sara", "pages": 420}
book3 = {"title": "AI Fundamentals", "author": "Omar", "pages": 300}

library = [book1, book2, book3]
genres = {"Programming", "Data Science"}

# 1
print(book1["title"])
print(book2["title"])
print(book3["title"])

# 2
book1["genre"] = "Programming"
book2["genre"] = "Data Science"
book3["genre"] = "Programming"

# 3
book3["pages"] = 320

# 4
book2.pop("author")

# 5
genres.add("Machine Learning")

# 6
genres.remove("Data Science")

# 7
print(library)
print(genres)
```
