المهمة 1 – التعامل مع قائمة الدرجات

قائمة الدرجات:
grades = [75, 88, 92, 60, 45, 100, 82, 90, 73, 55]

احسب عدد الطلاب.
1.	اعرض أعلى وأقل درجة
2.	احسب مجموع الدرجات والمتوسط
3.	أضف درجة جديدة 77 إلى القائمة
4.	احذف الدرجة 45 من القائمة
5.	رتب الدرجات ترتيبًا تصاعديًا
6.	اعرض الطلاب الناجحين فقط (درجة ≥ 60)
7.	عكس ترتيب الدرجات
8.	احسب عدد مرات تكرار الدرجة 90
9.	استخرج الدرجات من الفهرس 2 إلى 6
10.	أنشئ قائمتين جديدتين وادمجهما ثم كررهما مرتين

</div>

<div dir="rtl" style="text-align: right;">
الحل</div>

```python

grades = [75, 88, 92, 60, 45, 100, 82, 90, 73, 55]

# 1.Number of students
print("Number of students:", len(grades))

# 2.Highest and lowest grade
print("Highest grade:", max(grades))
print("Lowest grade:", min(grades))

# 3.Total and average
total = sum(grades)
avg = total / len(grades)
print("Total grades:", total)
print("Average:", avg)

# 4.Add a new grade
grades.append(77)
print("After adding:", grades)

# 5.Remove a grade
grades.remove(45)
print("After removing 45:", grades)

# 6.Sort in ascending order
grades.sort()
print("Ascending order:", grades)

# 7.Students who passed ( 60)
passed = [g for g in grades if g >= 60]
print("Passed students:", passed)

# 8.Reverse the list
grades.reverse()
print("Reversed order:", grades)

# 9.Count occurrences of 90
print("Number of times 90 appears:", grades.count(90))

# 10.Grades from index 2 to 6
print("Grades from index 2 to 6:", grades[2:7])

# 11.Merge two lists and repeat
list1 = ["Python", "Java"]
list2 = ["C++", "JavaScript"]
merged = list1 + list2
print("Merged list:", merged)

```

<div dir="rtl" style="text-align: right;">  المهمة 2 – إدارة منتجات في متجر

قائمة المنتجات:
products = ["Laptop", "Phone", "Tablet", "Camera", "Smartwatch"]

أسئلة التمرين:
1.	احسب عدد المنتجات.
2.	أضف منتجين جديدين: "Headphones" و "Monitor".
3.	احذف المنتج "Camera".
4.	رتب المنتجات أبجديًا.
5.	اعكس ترتيب المنتجات.
6.	تحقق مما إذا كان المنتج "Phone" موجودًا.
7.	أوجد فهرس المنتج "Laptop".
8.	أضف المنتج "Keyboard" في الفهرس 2.
9.	أنشئ قائمة فرعية من أول 3 منتجات.
10.	ادمج القائمة الفرعية مع القائمة الأصلية.

</div>

```python

products = ["Laptop", "Phone", "Tablet", "Camera", "Smartwatch"]

# 1.Print the number of products
print("Number of products:", len(products))

# 2.Add two new products
products.extend(["Headphones", "Monitor"])
print("After adding:", products)

# 3.Remove the product "Camera"
products.remove("Camera")
print("After removing Camera:", products)

# 4.Sort products alphabetically
products.sort()
print("Sorted:", products)

# 5.Reverse the list
products.reverse()
print("Reversed:", products)

# 6.Check if "Phone" exists
print("Is Phone present?", "Phone" in products)

# 7.Find the index of "Laptop"
print("Index of Laptop:", products.index("Laptop"))

# 8.Insert a product at a specific position (index 2)
products.insert(2, "Keyboard")
print("After inserting Keyboard:", products)

# 9.Create a sublist of the first 3 items
sub_list = products[:3]
print("Sublist:", sub_list)

# 10.Merge the sublist with the original list
combined = products + sub_list
print("Combined list:", combined)

```

المهمة 3 – تحليل مبيعات أسبوعية

قائمة المبيعات الأسبوعية:
sales = [120, 150, 90, 300, 500, 200, 250]

1.	احسب عدد الأيام.
2.	اعرض أعلى وأقل مبيعات.
3.	احسب مجموع المبيعات ومتوسطها.
4.	أضف مبيعات يوم جديد بقيمة 400.
5.	احذف مبيعات اليوم الثالث.
6.	رتب المبيعات ترتيبًا تصاعديًا ثم عكسيًا.
7.	احسب عدد مرات تكرار المبيعات 200.
8.	اعرض أول وآخر عنصر في القائمة.

</div>

```python

sales = [120, 150, 90, 300, 500, 200, 250]

# 1.Number of days
print("Number of days:", len(sales))

# 2.Highest and lowest sales
print("Highest sales:", max(sales))
print("Lowest sales:", min(sales))

# 3.Total and average sales
total = sum(sales)
average = total / len(sales)
print("Total sales:", total)
print("Average sales:", average)

# 4.Add sales for a new day
sales.append(400)
print("After adding new day:", sales)

# 5.Delete sales of the third day (index 2)
del sales[2]
print("After deleting third day:", sales)

# 6.Sort ascending then reverse
sales.sort()
print("Ascending order:", sales)
sales.reverse()
print("Descending order:", sales)

# 7.Count occurrences of 200
print("Number of times 200 appears:", sales.count(200))

# 8.First and last element
print("First:", sales[0], "Last:", sales[-1])

```

المهمة 4 – معالجة بيانات درجات نصية (متقدم)

قائمة البيانات:

numeric = [90, 85, 70, 100, 60, 95]

أسئلة التمرين:
1.	أضف درجة جديدة 80.
2.	احسب المجموع والمتوسط.
3.	تحقق من وجود الدرجة 100.
4.	رتب القيم ترتيبًا تنازليًا.
5.	استخرج أعلى 3 درجات.


</div>

<div dir="rtl" style="text-align: right;">  


</div>

```python

numeric = [90, 85, 70, 100, 60, 95]

# 1.Add a new grade 80
numeric.append(80)
print("After adding 80:", numeric)

# 2.Calculate total and average
total = sum(numeric)
avg = total / len(numeric)
print("Total:", total, "| Average:", avg)

# 3.Check if 100 exists
print("Is 100 present?", 100 in numeric)

# 4.Sort in descending order
numeric.sort(reverse=True)
print("Descending order:", numeric)

# 5.Extract top 3 grades
top3 = numeric[:3]
print("Top 3 grades:", top3)

```
