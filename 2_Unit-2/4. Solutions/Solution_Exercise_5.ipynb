التمرين 1 – بيانات منتج في متجر إلكتروني

product = {
    "name": "Laptop",
    "price": 1200,
    "brand": "Dell",
    "quantity": 10
}

المطلوب:

1.	اطبع اسم المنتج وسعره.
2.	عدّل الكمية لتصبح 15.
3.	أضف مفتاحًا جديدًا "discount" بقيمة 10 (تعني %10).
4.	احسب السعر بعد الخصم وأضفه في مفتاح "final_price".
5.	احذف المفتاح "brand".
6.	اطبع القاموس بعد التعديل.

الحل:




</div>

```python
product = {
        "name": "Laptop",
            "price": 1200,
                "brand": "Dell",
                    "quantity": 10
}
print(product["name"], product["price"])
product["quantity"] = 15
product["discount"] = 10
product["final_price"] = product["price"] - (product["price"] * product["discount"] / 100)
product.pop("brand")
print(product)
```

التمرين 2 – بيانات طالب

student = {
    "name": "Sara",
    "age": 22,
    "major": "Computer Science",
    "grades": {"math": 85, "python": 90, "databases": 88}
}

المطلوب:
1.	اطبع اسم الطالبة وتخصصها.
2.	أضف درجة جديدة للمادة "AI" بقيمة 95 داخل القاموس الفرعي "grades".
3.	عدّل درجة "math" لتصبح 90.
4.	احسب متوسط الدرجات وأضفه كمفتاح جديد "average".
5.	احذف مفتاح "age".
6.	اطبع القاموس بعد التعديل.
الحل:


</div>


```python
student = {
        "name": "Sara",
            "age": 22,
                "major": "Computer Science",
                    "grades": {"math": 85, "python": 90, "databases": 88}
}
print(student["name"], student["major"])
student["grades"]["AI"] = 95
student["grades"]["math"] = 90
student["average"] = sum(student["grades"].values()) / len(student["grades"])
student.pop("age")
print(student)
```

التمرين 4 – سجل حضور موظف

employee = {
    "name": "Layla",
    "position": "HR Manager",
    "salary": 7000,
    "attendance": {"Jan": 20, "Feb": 18, "Mar": 22}
}

المطلوب:

1.	اطبع اسم الموظفة ومسمّاها الوظيفي.
2.	أضف شهرًا جديدًا في سجل الحضور "Apr" بقيمة 21.
3.	عدّل عدد أيام الحضور في "Feb" لتصبح 20.
4.	احسب متوسط الحضور خلال الأشهر الأربعة وأضفه في مفتاح جديد "average_attendance".
5.	أضف علاوة جديدة "bonus" بقيمة 500 إلى القاموس.
6.	احذف المفتاح "position".
7.	اطبع القاموس بعد جميع التعديلات.

الحل



</div>

```python
employee = {"name": "Layla","position": "HR Manager","salary": 7000,"attendance": {"Jan": 20, "Feb": 18, "Mar": 22}}
# 1
print(employee["name"], employee["position"])
# 2
employee["attendance"]["Apr"] = 21
# 3
employee["attendance"]["Feb"] = 20
# 4
employee["average_attendance"] = sum(employee["attendance"].values()) / len(employee["attendance"])
# 5
employee["bonus"] = 500
# 6
employee.pop("position")
# 7
print(employee)
```
