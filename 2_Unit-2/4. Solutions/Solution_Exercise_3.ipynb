التمرين 1 – معالجة أسماء الطلاب

قائمة الطلاب:
students = ["Ali", "Sara", "Omar", "Lina", "Ali", "Yasmin"]

أسئلة التمرين:
1.	احسب عدد الطلاب.
2.	احسب عدد مرات تكرار الاسم "Ali".
3.	أضف اسمًا جديدًا "Khaled".
4.	احذف الاسم "Lina".
5.	رتب القائمة أبجديًا.
6.	عكس ترتيب القائمة.
7.	تحقق مما إذا كان الاسم "Omar" موجودًا.
8.	اعرض أول وآخر اسم في القائمة.
9.	استخرج أول 3 أسماء فقط.
10.	ادمج القائمة مع قائمة جديدة ["Nour", "Sami"].




</div>

```python
students = ["Ali", "Sara", "Omar", "Lina", "Ali", "Yasmin"]
# 1. Number of students
print("Number of students:", len(students))
# 2. Count how many times "Ali" appears
print("Occurrences of 'Ali':", students.count("Ali"))
# 3. Add a new student
students.append("Khaled")
# 4. Remove the student "Lina"
students.remove("Lina")
# 5. Sort the list alphabetically
students.sort()
# 6. Reverse the order
students.reverse()
# 7. Check if "Omar" is in the list
print("Is 'Omar' in the list?", "Omar" in students)
# 8. Show first and last student
print("First:", students[0], "Last:", students[-1])
# 9. Extract the first 3 students
print("First 3 students:", students[:3])
# 10. Merge with a new list of students
new_students = ["Nour", "Sami"]
combined = students + new_students
print("Combined list:", combined)
```

التمرين 2 – درجات متعددة المواد

قوائم الدرجات:

math = [90, 80, 70, 85, 95]
science = [88, 76, 92, 100, 67]

أسئلة التمرين:
1.	اطبع متوسط كل مادة.
2.	ادمج القائمتين في قائمة واحدة.
3.	احسب المتوسط العام لجميع الدرجات.
4.	اعرض درجات النجاح فقط (≥ 70).
5.	احذف أقل درجة.
6.	رتب الدرجات ترتيبًا تنازليًا.
7.	أضف درجة جديدة 77.
8.	استخرج الدرجات من الفهرس 2 إلى 5.

</div>


```python
math = [90, 80, 70, 85, 95]
science = [88, 76, 92, 100, 67]
# 1. Print the average for each subject
print("Math average:", sum(math) / len(math))
print("Science average:", sum(science) / len(science))
# 2. Combine both lists into one
all_grades = math + science
print("All grades:", all_grades)
# 3. Calculate overall average
overall_avg = sum(all_grades) / len(all_grades)
print("Overall average:", overall_avg)
# 4. Only passing grades (>= 70)
passed = [g for g in all_grades if g >= 70]
print("Passing grades:", passed)
# 5. Remove the lowest grade
all_grades.remove(min(all_grades))
print("After removing the lowest grade:", all_grades)
# 6. Sort in descending order
all_grades.sort(reverse=True)
print("Grades in descending order:", all_grades)
# 7. Add a new grade
all_grades.append(77)
# 8. Extract grades from index 2 to 5
print("Partial grades:", all_grades[2:6])
```
