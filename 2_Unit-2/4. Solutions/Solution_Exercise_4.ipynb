التمرين 1: درجات الطلاب في مواد متعددة

grades = [
    [90, 85, 88],
    [70, 75, 80],
    [100, 95, 98],
    [60, 55, 70]
]
المطلوب:
1.	احسب عدد الطلاب.
2.	احسب مجموع الدرجات لكل طالب (قائمة جديدة).
3.	احسب المتوسط العام لجميع الدرجات في الصف.
4.	أوجد أعلى درجة بين جميع الطلاب.
5.	اجمع الدرجات في قائمة واحدة فقط (بدون الحلقات).

الحل




</div>

```python
grades = [
        [90, 85, 88],
            [70, 75, 80],
                [100, 95, 98],
                    [60, 55, 70]
]
# 1
num_students = len(grades)
# 2
total_per_student = [sum(grades[0]), sum(grades[1]), sum(grades[2]), sum(grades[3])]
# 3
overall_average = (sum(sum(grades, []))) / (len(sum(grades, [])))
# 4
highest_grade = max(sum(grades, []))
# 5
flat_list = sum(grades, [])
print(num_students)
print(total_per_student)
print(overall_average)
print(highest_grade)
print(flat_list)
```

التمرين 2: أسعار منتجات في فروع مختلفة

prices = [
    [20, 25, 22],
    [18, 30, 27],
    [21, 19, 23]
]

المطلوب:
1.	احسب متوسط الأسعار في كل فرع.
2.	أوجد أقل سعر من بين كل الفروع.
3.	اجمع الأسعار في قائمة واحدة.
4.	احسب مجموع أسعار الأعمدة (المنتجات نفسها في كل الفروع).
5.	استخرج أعلى سعر في كل فرع.
الحل

</div>


```python
prices = [
        [20, 25, 22],
            [18, 30, 27],
                [21, 19, 23]
]
# 1
avg_per_branch = [sum(prices[0])/len(prices[0]), sum(prices[1])/len(prices[1]), sum(prices[2])/len(prices[2])]
# 2
lowest_price = min(sum(prices, []))
# 3
all_prices = sum(prices, [])
# 4
total_per_product = [prices[0][0]+prices[1][0]+prices[2][0],
                                          prices[0][1]+prices[1][1]+prices[2][1],
                                                               prices[0][2]+prices[1][2]+prices[2][2]]
# 5
max_per_branch = [max(prices[0]), max(prices[1]), max(prices[2])]
print(avg_per_branch)
print(lowest_price)
print(all_prices)
print(total_per_product)
print(max_per_branch)
```
