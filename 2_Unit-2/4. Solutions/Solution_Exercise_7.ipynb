التمرين1– الفواكه المفضلة.

fruits1 = {"Apple", "Banana", "Orange"}

fruits2 = {"Banana", "Grapes", "Mango"}

المطلوب:

1.	أضف "Pineapple" إلى المجموعة الأولى.
2.	احذف "Orange" من المجموعة الأولى.
3.	اطبع اتحاد المجموعتين.
4.	اطبع العناصر المشتركة فقط.
5.	اطبع الفواكه الموجودة في الأولى وليست في الثانية.
6.	اطبع الفواكه الفريدة لكل مجموعة (الفرق المتماثل).
الحل:




</div>

```python
fruits1 = {"Apple", "Banana", "Orange"}
fruits2 = {"Banana", "Grapes", "Mango"}
fruits1.add("Pineapple")
fruits1.remove("Orange")
print("Union:", fruits1 | fruits2)
print("Intersection:", fruits1 & fruits2)
print("Difference (fruits1 - fruits2):", fruits1 - fruits2)
print("Symmetric Difference:", fruits1 ^ fruits2)
```

التمرين 2 – المدن التي زارها شخصان

visited1 = {"Paris", "London", "Berlin"}

visited2 = {"Rome", "Paris", "Madrid"}

المطلوب:

1.	أضف "Tokyo" إلى المجموعة الأولى.
2.	احذف "Berlin" من المجموعة الأولى.
3.	اطبع جميع المدن التي زارها أي منهما.
4.	اطبع المدن المشتركة بينهما.
5.	اطبع المدن التي زارها الأول فقط.
6.	اطبع المدن التي زارها أحدهما دون الآخر (الفرق المتماثل).
الحل:


</div>


```python
visited1 = {"Paris", "London", "Berlin"}
visited2 = {"Rome", "Paris", "Madrid"}
visited1.add("Tokyo")
visited1.remove("Berlin")
print("Union:", visited1 | visited2)
print("Intersection:", visited1 & visited2)
print("Difference (visited1 - visited2):", visited1 - visited2)
print("Symmetric Difference:", visited1 ^ visited2)
```
