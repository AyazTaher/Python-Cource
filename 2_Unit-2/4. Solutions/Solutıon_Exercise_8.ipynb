التمرين 4 – رحلة سفر

trip1 = ["Paris", "London", "Berlin"]

trip2 = ["Rome", "Paris", "Madrid"]

visited = set(trip1)

planned = set(trip2)

المطلوب:
1.	أضف "Tokyo" و "Dubai" إلى مجموعة visited.
2.	احذف "Berlin" من visited.
3.	اطبع اتحاد المجموعتين.
4.	اطبع التقاطع بين المجموعتين.
5.	اطبع المدن الموجودة في visited فقط.
6.	اطبع المدن غير المشتركة (الفرق المتماثل).
الحل:




</div>

```python
trip1 = ["Paris", "London", "Berlin"]
trip2 = ["Rome", "Paris", "Madrid"]

visited = set(trip1)
planned = set(trip2)



visited.add("Tokyo")
visited.add("Dubai")

visited.remove("Berlin")

print("Union:", visited | planned)
print("Intersection:", visited & planned)
print("Difference (visited - planned):", visited - planned)
print("Symmetric Difference:", visited ^ planned)

```
