المهمة 1-  المهارات المكتسبة

skills1 = {"Python", "SQL", "Excel"}

skills2 = {"SQL", "PowerBI", "Python"}

المطلوب:
1.	أضف "R" إلى المهارات الخاصة بالمجموعة الأولى.
2.	احذف "Excel" من المجموعة الأولى.
3.	اطبع كل المهارات الموجودة في أي من المجموعتين.
4.	اطبع المهارات المشتركة فقط.
5.	اطبع المهارات التي يمتلكها الشخص الأول فقط.
6.	اطبع المهارات الفريدة لكل شخص (الفرق المتماثل).
الحل:




</div>

<div dir="rtl" style="text-align: right;">  


</div>

```python


skills1 = {"Python", "SQL", "Excel"}
skills2 = {"SQL", "PowerBI", "Python"}
#1
skills1.add("R")
#2
skills1.remove("Excel")
#3
print("Union:", skills1 | skills2)
#4
print("Intersection:", skills1 & skills2)
#5
print("Difference (skills1 - skills2):", skills1 - skills2)
#6
print("Symmetric Difference:", skills1 ^ skills2)

```

المهمة 2 – المهارات التقنية لفريق

team1 = {"Python", "JavaScript", "SQL", "Git"}

team2 = {"Python", "Java", "Docker", "SQL"}

المطلوب:
1.	أضف "HTML" و "CSS" إلى مجموعة الفريق الأول.
2.	احذف "Git" من الفريق الأول.
3.	اطبع كل المهارات الموجودة في أي من الفريقين (الاتحاد).
4.	اطبع المهارات المشتركة بين الفريقين (التقاطع).
5.	اطبع المهارات التي يمتلكها الفريق الأول فقط (الفرق).
6.	اطبع المهارات الفريدة لكل فريق (الفرق المتماثل).
7.	تحقق إذا كان "Docker" موجودًا في الفريق الأول.
الحل:



</div>

```python

team1 = {"Python", "JavaScript", "SQL", "Git"}
team2 = {"Python", "Java", "Docker", "SQL"}

# 1
team1.update(["HTML", "CSS"])
# 2
team1.remove("Git")
# 3
print("Union:", team1 | team2)
# 4
print("Intersection:", team1 & team2)
# 5
print("Difference (team1 - team2):", team1 - team2)
# 6
print("Symmetric Difference:", team1 ^ team2)
# 7
print("Is 'Docker' in team1?", "Docker" in team1)
```
