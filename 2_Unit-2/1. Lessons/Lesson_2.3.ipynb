<div dir="rtl" style="text-align: right;">
2-	الوصول إلى القيم</div>

```python

person = {"name": "Sara", "age": 22}

print(person["name"])

print(person["age"]) 




```

<div dir="rtl" style="text-align: right;">
3-	تعديل القيم</div>

```python

person = {"name": "Sara", "age": 22}

print(person)

person = {"name": "Mona"}
person["city"] = "Damascus"

print(person)




```

<div dir="rtl" style="text-align: right;">
4-	إضافة عنصر جديد</div>

```python

person = {"name": "Omar", "age": 28}
print(person)

person = {"name": "Mona"}
person = {"age":"30"}

print(person)




```

<div dir="rtl" style="text-align: right;">
5-	حذف عنصر</div>

```python

person = {"name": "Omar", "age": 28}
print(person)

del person["age"]
print(person)



```

التمرين التطبيقي
لديك قاموس يمثل بيانات موظف

employee = {
    "name": "Ali",
    "age": 30,
    "city": "Aleppo",
    "salary": 5000}
    
المطلوب:
1.	اطبع اسم الموظف وعمره
2.	عدّل قيمة الراتب إلى 6000
3.	أضف مفتاح جديد اسمه "department" وقيمته "IT"
4.	احذف المفتاح "city"
5.	اطبع القاموس بعد التعديلات

الحل</div>

```python

employee = {
        "name": "Ali",
            "age": 30,
                "city": "Aleppo",
                    "salary": 5000
}

print(employee)


# 1
print("Name:", employee["name"])
print("Age:", employee["age"])

employee = {
        "name": "Ali",
            "age": 30,
                "city": "Aleppo",
                    "salary": 5000
}
print(employee)

employee["salary"] = 6000
print(employee)


employee = {
        "name": "Ali",
            "age": 30,
                "city": "Aleppo",
                    "salary": 5000
}
print(employee)

employee["department"] = "IT"
print(employee)


employee = {
        "name": "Ali",
            "age": 30,
                "city": "Aleppo",
                    "salary": 5000
}
print(employee)
del employee["city"]
print(employee)


```
