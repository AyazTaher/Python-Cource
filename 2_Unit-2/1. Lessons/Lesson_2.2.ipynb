<div dir="rtl" style="text-align: right;">
1.	إنشاء القوائم المتداخلة</div>

```python

# 1. Create nested list
nested_list = [
        ["Ali", 90, "Damascus"],
            ["Sara", 85, "Aleppo"],
                ["Omar", 78, "Homs"]
]

```

<div dir="rtl" style="text-align: right;">
2.	الوصول إلى عناصر القوائم المتداخلة</div>

```python

# 2. Access elements

# Access Sara

print(nested_list[1][0])

# Access city Homs

print(nested_list[2][2])



```

<div dir="rtl" style="text-align: right;">
3.	تعديل القوائم المتداخلة</div>

```python

# 3. Modify nested list

nested_list = [
        ["Ali", 90, "Damascus"],
            ["Sara", 85, "Aleppo"],
                ["Omar", 78, "Homs"]
]

nested_list[1][1] = 95

print(nested_list)



```

<div dir="rtl" style="text-align: right;">
4-	إنشاء جدول بيانات بسيط</div>

```python



# 4. Create a simple data table

students = [
        ["Ali", 90, "Math"],
            ["Sara", 85, "Physics"],
                ["Omar", 78, "Chemistry"]
]

print(students[0])
print(students[1][2])

```

<div dir="rtl" style="text-align: right;">
5-	 تعديل بيانات داخل قائمة متداخلة</div>

```python

# 5. Modify data inside nested list

students = [
        ["Ali", 90, "Math"],
            ["Sara", 85, "Physics"],
                ["Omar", 78, "Chemistry"]
]

students[1][1] = 92
students[2][2] = "Biology"

print(students)


```
