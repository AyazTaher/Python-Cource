<div dir="rtl" style="text-align: right;">
التمرين التطبيقبي
لديك قائمة تمثل درجات طلاب في مادة الرياضيات

grades = [75, 88, 92, 60, 45, 100, 82, 90, 73, 55]

المطلوب
1.	اطبع عدد الطلاب
2.	أوجد أعلى درجة وأقل درجة
3.	احسب مجموع الدرجات والمتوسط الحسابي
4.	أضف درجة جديدة (مثل: 77) إلى القائمة
5.	احذف الدرجة (45) من القائمة
6.	رتّب الدرجات تصاعديًا
7.	استخرج قائمة جديدة تحتوي فقط على الطلاب الذين نجحوا (الدرجة ≥ 60)
8.	اعكس ترتيب القائمة
9.	أوجد عدد مرات تكرار درجة 90
10.	استخرج الدرجات من الفهرس 2 حتى 6

الحل

</div>


```python
grades = [75, 88, 92, 60, 45, 100, 82, 90, 73, 55]

# 1 - Number of students
print("Number of students:", len(grades))


# 2 - Maximum and minimum grades
print("Highest grade:", max(grades))
print("Lowest grade:", min(grades))

# 3 - Sum and average
print("Sum of grades:", sum(grades))
print("Average grade:", sum(grades) / len(grades))

# 4 - Add a grade
grades.append(77)
print("After adding:", grades)

# 5 - Remove a grade
grades.remove(45)
print("After removing:", grades)

# 6 - Sort grades ascending
grades.sort()
print("Sorted ascending:", grades)

# 7 - Passed students (grades >= 60)
passed = [g for g in grades if g >= 60]
print("Passed students:", passed)

# 8 - Reverse grades
grades.reverse()
print("Reversed grades:", grades)

# 9 - Count occurrences of 90
print("90 occurred:", grades.count(90), "times")

# 10 - Slice grades from index 2 to 6
print("Slice from index 2 to 6:", grades[2:7])
```
