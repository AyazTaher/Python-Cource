<div dir="rtl" style="text-align: right;">
3-	العمليات على الـ Set</div>

<div dir="rtl" style="text-align: right;">
1-	الإضافة</div>

```python

fruits = {"Apple", "Banana"}
# add()

fruits.add("Orange")
print(fruits)

# update ()
fruits.update(["Grapes", "Watermelon"])
print(fruits)  




```

<div dir="rtl" style="text-align: right;">
2-	الحذف</div>

```python

cities = {"London", "Paris", "Berlin"}
# remove()
cities.remove("London")
print(cities)

# discard()
cities.discard("Rome")
print(cities)


# pop()
removed_item = cities.pop()
print("Removed",removed_item )
# clear()
cities.clear()
print(cities)




```

<div dir="rtl" style="text-align: right;">
3-	 دعم العمليات الرياضية</div>

<div dir="rtl" style="text-align: right;">
-	الاتحاد (Union)</div>

```python

set1 = {1, 2, 3}
set2 = {3, 4, 5}

print(set1 | set2)
print(set1.union(set2)) 



```

<div dir="rtl" style="text-align: right;">
-	التقاطع (Intersection)</div>

```python

set1 = {1, 2, 3}
set2 = {2, 3, 4}

print(set1 & set2)
print(set1.intersection(set2))


```

<div dir="rtl" style="text-align: right;">
-	الفرق (Difference)</div>

```python

set1 = {1, 2, 3}
set2 = {2, 3, 4}

print(set1 - set2)
print(set1.difference(set2))


```

<div dir="rtl" style="text-align: right;">
-	الفرق المتماثل (Symmetric Difference)</div>

```python

set1 = {1, 2, 3}
set2 = {2, 3, 4}

print(set1 ^ set2)
print(set1.symmetric_difference(set2))



```
