الوحدة الثانية: تراكيب البيانات المتقدمة
</div>

<div dir="rtl" style="text-align: right;">
مشروع تطبيقي مبسط: بيانات الطلاب
</div>

<div dir="rtl" style="text-align: right;">
الخطوة 1: إعداد البيانات
</div>

<div dir="rtl" style="text-align: right;">
-  قائمة أسماء الطلاب
</div>

```python

students = ["Ali", "Sara", "Mona"]
print(students) 

```

<div dir="rtl" style="text-align: right;">
- بيانات الطلاب: [العمر، المدينة، المعدل]
</div>

```python

students_data = [
        [21, "Damascus", 85],
            [22, "Aleppo", 90],
                [20, "Homs", 78]
]

print(students_data) 
```

<div dir="rtl" style="text-align: right;">
- المواد الدراسية لكل طالب (Nested List)
</div>

```python

subjects = [
        ["Math", "Physics", "Python"],
            ["Math", "Chemistry", "Python"],
                ["Biology", "Python", "Chemistry"]
]
print(subjects) 
```

<div dir="rtl" style="text-align: right;">
- هوايات الطلاب (Set)
</div>

```python

hobbies_ali = {"Football", "Reading"}
hobbies_sara = {"Reading", "Drawing"}
hobbies_mona = {"Football", "Drawing"}

print(hobbies_ali) 
print(hobbies_sara) 
print(hobbies_mona) 
```

<div dir="rtl" style="text-align: right;">
- بيانات ثابتة لكل طالب (Tuple): (الجنس، سنة الميلاد)
</div>

```python

info_ali = ("Male", 2002)
info_sara = ("Female", 2001)
info_mona = ("Female", 2003)

print(info_ali)
print(info_sara)
print(info_mona)

```

<div dir="rtl" style="text-align: right;">
الخطوة 2: استخدام القوائم المتداخلة</div>

<div dir="rtl" style="text-align: right;">
- الوصول إلى بيانات طالب معين</div>

```python

print(students[0]) 
print(students_data[1][2])
print(subjects[2][0])
```

<div dir="rtl" style="text-align: right;">
الخطوة 3: استخدام الـ Dictionary</div>

<div dir="rtl" style="text-align: right;">
- إنشاء قاموس طالب واحد</div>

```python

ali_record = {
        "age": students_data[0][0],
            "city": students_data[0][1],
                "grade": students_data[0][2],
                    "subjects": subjects[0],
                        "hobbies": hobbies_ali,
                            "info": info_ali
}

print(ali_record["city"]) 
print(ali_record["subjects"])
print(ali_record["info"])
```

<div dir="rtl" style="text-align: right;">
الخطوة 4: استخدام الـ Tuple</div>

<div dir="rtl" style="text-align: right;">
- الوصول إلى الجنس وسنة الميلاد</div>

```python


print(info_ali[0]) 
print(info_ali[1]) 
```

<div dir="rtl" style="text-align: right;">
الخطوة 5: استخدام الـ Set</div>

<div dir="rtl" style="text-align: right;">
- إضافة هواية جديدة</div>

```python


hobbies_ali = {"Football", "Reading"}
print(hobbies_ali) 

hobbies_ali.add("Music")
print(hobbies_ali) 
```

<div dir="rtl" style="text-align: right;">
- إزالة هواية</div>

```python

hobbies_ali = {"Football", "Reading"}
print(hobbies_ali)

hobbies_ali.remove("Football")
print(hobbies_ali)
```

<div dir="rtl" style="text-align: right;">
-  اتحاد الهوايات بين Ali و Sara</div>

```python

hobbies_ali = {"Football", "Reading"}
hobbies_sara = {"Reading", "Drawing"}

all_hobbies = hobbies_ali | hobbies_sara
print(all_hobbies) 
```

<div dir="rtl" style="text-align: right;">
الخطوة 6: دمج البيانات البسيط</div>

<div dir="rtl" style="text-align: right;">
-  دمج المواد الدراسية لAli وSara</div>

```python

combined_subjects = subjects[0] + subjects[1]
print(combined_subjects) 
```

<div dir="rtl" style="text-align: right;">
ملخص العمليات

- List: تخزين بيانات قابلة للتغيير والوصول إليها بالفهارس

- Nested List: تخزين بيانات متعددة لكل طالب

- Dictionary: تخزين بيانات كل طالب بطريقة مفتاح: قيمة

- Tuple: تخزين بيانات ثابتة لا تتغير

- Set: إضافة، حذف، اتحاد، تقاطع لإدارة البيانات الفريدة</div>
