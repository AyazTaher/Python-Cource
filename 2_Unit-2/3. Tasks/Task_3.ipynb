المهمة 1 – التعامل مع قائمة الدرجات

قائمة الدرجات:
grades = [75, 88, 92, 60, 45, 100, 82, 90, 73, 55]

احسب عدد الطلاب.
1.	اعرض أعلى وأقل درجة
2.	احسب مجموع الدرجات والمتوسط
3.	أضف درجة جديدة 77 إلى القائمة
4.	احذف الدرجة 45 من القائمة
5.	رتب الدرجات ترتيبًا تصاعديًا
6.	اعرض الطلاب الناجحين فقط (درجة ≥ 60)
7.	عكس ترتيب الدرجات
8.	احسب عدد مرات تكرار الدرجة 90
9.	استخرج الدرجات من الفهرس 2 إلى 6
10.	أنشئ قائمتين جديدتين وادمجهما ثم كررهما مرتين

</div>

<div dir="rtl" style="text-align: right;">
الحل</div>

```python
grades = [75, 88, 92, 60, 45, 100, 82, 90, 73, 55]

# 1. عدد الطلاب
print("عدد الطلاب:", len(grades))

# 2. أعلى وأقل درجة
print("أعلى درجة:", max(grades))
print("أقل درجة:", min(grades))

# 3. مجموع الدرجات والمتوسط
total = sum(grades)
avg = total / len(grades)
print("مجموع الدرجات:", total)
print("المتوسط:", avg)

# 4. إضافة درجة جديدة
grades.append(77)
print("بعد الإضافة:", grades)

# 5. حذف درجة معينة
grades.remove(45)
print("بعد الحذف:", grades)

# 6. ترتيب تصاعدي
grades.sort()
print("الترتيب التصاعدي:", grades)

# 7. الطلاب الناجحون (≥ 60)
passed = [g for g in grades if g >= 60]
print("الطلاب الناجحون:", passed)

# 8. عكس الترتيب
grades.reverse()
print("الترتيب العكسي:", grades)

# 9. عدد مرات تكرار درجة 90
print("عدد مرات ظهور 90:", grades.count(90))

# 10. الدرجات من الفهرس 2 إلى 6
print("درجات الفهرس 2 إلى 6:", grades[2:7])

# 11. قائمتين ولصقهما وتكرارهما
list1 = ["Python", "Java"]
list2 = ["C++", "JavaScript"]
merged = list1 + list2
repeated = merged * 2
print("القائمة المدمجة:", merged)
print("بعد التكرار:", repeated)
```

<div dir="rtl" style="text-align: right;">  المهمة 2 – إدارة منتجات في متجر

قائمة المنتجات:
products = ["Laptop", "Phone", "Tablet", "Camera", "Smartwatch"]

أسئلة التمرين:
1.	احسب عدد المنتجات.
2.	أضف منتجين جديدين: "Headphones" و "Monitor".
3.	احذف المنتج "Camera".
4.	رتب المنتجات أبجديًا.
5.	اعكس ترتيب المنتجات.
6.	تحقق مما إذا كان المنتج "Phone" موجودًا.
7.	أوجد فهرس المنتج "Laptop".
8.	أضف المنتج "Keyboard" في الفهرس 2.
9.	أنشئ قائمة فرعية من أول 3 منتجات.
10.	ادمج القائمة الفرعية مع القائمة الأصلية.

</div>

```python
products = ["Laptop", "Phone", "Tablet", "Camera", "Smartwatch"]

# 1. اطبع عدد المنتجات
print("عدد المنتجات:", len(products))

# 2. أضف منتجين جديدين
products.extend(["Headphones", "Monitor"])
print("بعد الإضافة:", products)

# 3. احذف المنتج "Camera"
products.remove("Camera")

# 4. رتّب المنتجات أبجديًا
products.sort()
print("مرتبة:", products)

# 5. اعكس الترتيب
products.reverse()
print("عكسي:", products)

# 6. تحقق إن كان "Phone" موجودًا
print("هل يوجد Phone؟", "Phone" in products)

# 7. أوجد فهرس العنصر "Laptop"
print("فهرس Laptop:", products.index("Laptop"))

# 8. أضف المنتج في موقع محدد (في الفهرس 2)
products.insert(2, "Keyboard")
print("بعد الإدراج:", products)

# 9. أنشئ قائمة فرعية من أول 3 عناصر
sub_list = products[:3]
print("قائمة فرعية:", sub_list)

# 10. أعد دمجها مع القائمة الأصلية
combined = products + sub_list
print("قائمة مدمجة:", combined)
```

المهمة 3 – تحليل مبيعات أسبوعية

قائمة المبيعات الأسبوعية:
sales = [120, 150, 90, 300, 500, 200, 250]

1.	احسب عدد الأيام.
2.	اعرض أعلى وأقل مبيعات.
3.	احسب مجموع المبيعات ومتوسطها.
4.	أضف مبيعات يوم جديد بقيمة 400.
5.	احذف مبيعات اليوم الثالث.
6.	اعرض الأيام التي تجاوزت المبيعات فيها المتوسط.
7.	رتب المبيعات ترتيبًا تصاعديًا ثم عكسيًا.
8.	احسب عدد مرات تكرار المبيعات 200.
9.	اعرض أول وآخر عنصر في القائمة.
10.	أنشئ قائمة تحتوي على المبيعات مضاعفة.

</div>

```python
sales = [120, 150, 90, 300, 500, 200, 250]

# 1. عدد الأيام
print("عدد الأيام:", len(sales))

# 2. أعلى وأقل مبيعات
print("أعلى مبيعات:", max(sales))
print("أقل مبيعات:", min(sales))

# 3. مجموع المبيعات ومتوسطها
total = sum(sales)
average = total / len(sales)
print("المجموع:", total)
print("المتوسط:", average)

# 4. أضف مبيعات يوم جديد
sales.append(400)

# 5. احذف مبيعات اليوم الثالث
del sales[2]
print("بعد الحذف:", sales)

# 6. المبيعات فوق المتوسط فقط
above_avg = [x for x in sales if x > average]
print("أيام المبيعات العالية:", above_avg)

# 7. ترتيب تصاعدي ثم عكسي
sales.sort()
print("ترتيب تصاعدي:", sales)
sales.reverse()
print("ترتيب تنازلي:", sales)

# 8. عدد مرات تكرار 200
print("عدد مرات 200:", sales.count(200))

# 9. أول وآخر عنصر
print("أول:", sales[0], "آخر:", sales[-1])

# 10. إنشاء قائمة مضاعفة للقيم
double_sales = [x * 2 for x in sales]
print("المبيعات مضاعفة:", double_sales)
```

المهمة 6 – معالجة بيانات درجات نصية (متقدم)

قائمة البيانات:
data = ["90", "85", " seventy ", "100", " sixty", "95 "]
أسئلة التمرين:
1.	نظف النصوص من الفراغات.
2.	حول القيم الرقمية فقط إلى أعداد صحيحة.
3.	أضف درجة جديدة 80.
4.	احسب المجموع والمتوسط.
5.	تحقق من وجود الدرجة 100.
6.	رتب القيم ترتيبًا تنازليًا.
7.	استخرج أعلى 3 درجات.
8.	أنشئ قائمة بمضاعفة كل درجة.


</div>

<div dir="rtl" style="text-align: right;">  


</div>

```python
data = ["90", "85", " seventy ", "100", " sixty", "95 "]

# 1. تنظيف النصوص من الفراغات
cleaned = [x.strip() for x in data]
print("بعد التنظيف:", cleaned)

# 2. تحويل القيم الرقمية فقط إلى int
numeric = [int(x) for x in cleaned if x.isdigit()]
print("القيم الرقمية:", numeric)

# 3. إضافة درجة جديدة 80
numeric.append(80)

# 4. حساب المجموع والمتوسط
total = sum(numeric)
avg = total / len(numeric)
print("المجموع:", total, "| المتوسط:", avg)

# 5. التحقق من وجود 100
print("هل توجد درجة 100؟", 100 in numeric)

# 6. ترتيب تنازلي
numeric.sort(reverse=True)
print("الترتيب التنازلي:", numeric)

# 7. استخراج أعلى 3 درجات
top3 = numeric[:3]
print("أعلى 3:", top3)

# 8. مضاعفة كل درجة
double = [x * 2 for x in numeric]
print("مضاعفة:", double)

```
