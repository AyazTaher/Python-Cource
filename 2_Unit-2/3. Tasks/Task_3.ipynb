<div dir="rtl" style="text-align: right;">  المهمة 2 – إدارة منتجات في متجر

قائمة المنتجات:
products = ["Laptop", "Phone", "Tablet", "Camera", "Smartwatch"]

أسئلة التمرين:
1.	احسب عدد المنتجات.

3.	احذف المنتج "Camera".
4.	رتب المنتجات أبجديًا.
5.	اعكس ترتيب المنتجات.
6.	تحقق مما إذا كان المنتج "Phone" موجودًا.
7.	أوجد فهرس المنتج "Laptop".
8.	أضف المنتج "Keyboard" في الفهرس 2.
9.	أنشئ قائمة فرعية من أول 3 منتجات.
10.	ادمج القائمة الفرعية مع القائمة الأصلية.

المهمة 3 – تحليل مبيعات أسبوعية

قائمة المبيعات الأسبوعية:
sales = [120, 150, 90, 300, 500, 200, 250]

1.	احسب عدد الأيام.
2.	اعرض أعلى وأقل مبيعات.
3.	احسب مجموع المبيعات ومتوسطها.
4.	أضف مبيعات يوم جديد بقيمة 400.
5.	احذف مبيعات اليوم الثالث.
6.	اعرض الأيام التي تجاوزت المبيعات فيها المتوسط.
7.	رتب المبيعات ترتيبًا تصاعديًا ثم عكسيًا.
8.	احسب عدد مرات تكرار المبيعات 200.
9.	اعرض أول وآخر عنصر في القائمة.
10.	أنشئ قائمة تحتوي على المبيعات مضاعفة.
