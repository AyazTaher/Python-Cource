<div dir="rtl" style="text-align: right;">
1-	 استخدام and</div>

```python

age = 25
has_license = True

if age >= 18 and has_license:
        print("You can drive a car.")
else:
            print("You cannot drive.")
            
```

<div dir="rtl" style="text-align: right;">
2-	 استخدام or</div>

```python

is_weekend = False
is_holiday = False

if is_weekend or is_holiday:
        print("Today is a holiday!")
else:
            print("Today is a regular workday.")
            
```

<div dir="rtl" style="text-align: right;">
3-استخدام not</div>

```python

is_raining = True

if not is_raining:
        print("You can go outside, the weather is clear.")
else:
        print("It's better to stay at home.")
```

المطلوب

اكتب برنامجًا يطلب من المستخدم إدخال عمره، وحالته الصحية (نص: "yes" أو "no" للتطعيم)، ثم:
- إذا كان عمره أكبر من أو يساوي 18 و حالته الصحية تسمح، اطبع "مؤهل للتطعيم"
- إذا كان عمره أقل من 18 أو حالته الصحية لا تسمح، اطبع "غير مؤهل للتطعيم"
- استخدم أيضًا معامل not في أحد الشروط.

الحل 
</div>

```python

1 # Vaccination eligibility example

age = int(input("Enter your age: "))
health_ok = input("Is your health condition good for vaccination? (yes/no): ").lower()

if age >= 18 and health_ok == "yes":
        print("Eligible for vaccination.")
elif age < 18 or health_ok == "no":
            print("Not eligible for vaccination.")

# Using 'not' as an additional example

health_ok = input("Is your health condition good for vaccination? (yes/no): ").lower()
if not (health_ok == "yes"):
        print("Please consult a doctor before vaccination.")

```
