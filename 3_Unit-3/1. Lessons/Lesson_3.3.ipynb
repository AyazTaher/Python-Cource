الدرس 3.3 – عمليات المقارنة في بايثون</div>

<div dir="rtl" style="text-align: right;">
1-	 مقارنة الأرقام</div>

```python

x = 10
y = 20

print(x == y)
print(x != y)
print(x < y)
print(x >= 5)

```

<div dir="rtl" style="text-align: right;">
2-	 مقارنة النصوص</div>

```python

name1 = "Ali"
name2 = "Omar"

print(name1 == name2)
print(name1 < name2)

```

<div dir="rtl" style="text-align: right;">
3-	المقارنة المتسلسلة</div>

```python

age = 31
print(18 <= age <= 30)

```

<div dir="rtl" style="text-align: right;">
تمرين 
المطلوب

اكتب برنامجًا يطلب من المستخدم إدخال عمره ودرجته في اختبار (من 100)، ثم
-	إذا كان العمر بين 18 و30 والدرجة أكبر أو تساوي 50 → اطبع "ناجح"
-	إذا كانت الدرجة أقل من 50 → اطبع "راسب"
-	استخدم جميع معاملات المقارنة على الأقل مرة واحدة في البرنامج

الحل 
</div>

```python

age = int(input("Enter your age: "))
score = int(input("Enter your score: "))

if 18 <= age <= 30 and score >= 50:
        print("Pass")
elif score < 50:
            print("Fail")
else:
            print("Not eligible for classification")

            
```
