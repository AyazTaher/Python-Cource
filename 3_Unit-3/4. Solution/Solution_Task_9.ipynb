المهمة 1 – القوة والجذور
المطلوب:

اطلب من المستخدم إدخال عددين عشريين، ثم احسب واطبع:
1.	العدد الأول مرفوع للقوة الثانية.
2.	العدد الثاني مرفوع للقوة الثالثة.
4.	الجذر التكعيبي للعدد الثاني.
5.	مجموع القيم المرفوعة للقوى.
6.	الفرق بين العددين بعد رفعهما للقوى.
7.	ناتج الضرب بعد رفع العددين للقوى.

الحل:


</div>

```python

num1 = float(input("Enter the first number: "))
num2 = float(input("Enter the second number: "))

print(num1)
print(num2)
print("First number squared:", num1**2)
print("Second number cubed:", num2**3)
print("Cube root of the second number:", num2**(1/3))
print("Sum of powers:", num1**2 + num2**3)
print("Difference after powers:", num1**2 - num2**3)
print("Multiplication after powers:", num1**2 * num2**3)


```

المهمة 2 – خصائص العددين
المطلوب:

اطلب من المستخدم إدخال عددين صحيحين، ثم احسب واطبع:
1.	مجموع العددين.
2.	حاصل ضربهما.
3.	ناتج القسمة العشرية للعدد الأول على الثاني.
4.	مجموع العددين إذا كانا كلاهما زوجيين.
5.	تحقق إذا كان العددان كلاهما أكبر من 10.
6.	تحقق إذا كان العددان كلاهما سالبين.
7.	تحقق إذا كان العدد الأول مضاعف للعدد الثاني.
الحل:


</div>

```python

num1 = int(input("Enter the first number: "))
num2 = int(input("Enter the second number: "))

print(num1)
print(num2)
print("Sum:", num1 + num2)
print("Multiplication:", num1 * num2)
print("Decimal division:", num1 / num2)
print("Sum if both are even:", (num1 + num2) if (num1 % 2 == 0 and num2 % 2 == 0) else "Not available")
print("Are both greater than 10?", num1 > 10 and num2 > 10)
print("Are both negative?", num1 < 0 and num2 < 0)
print("Is the first a multiple of the second?", num1 % num2 == 0)
```
