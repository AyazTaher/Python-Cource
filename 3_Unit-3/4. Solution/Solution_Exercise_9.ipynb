التمرين 1 – العمليات الرياضية الأساسية
المطلوب:

اطلب من المستخدم إدخال عددين، ثم احسب واطبع:
1.	مجموع العددين.
2.	الفرق المطلق بين العددين.
3.	حاصل ضرب العددين.
4.	المتوسط الحسابي لهما.
5.	أكبر عدد بينهما.
6.	أصغر عدد بينهما.
7.	مجموع مربعي العددين.
الحل:


</div>

```python
num1 = float(input("Enter the first number: "))
num2 = float(input("Enter the second number: "))
print("Sum:", num1 + num2)
print("Absolute difference:", abs(num1 - num2))
print("Multiplication:", num1 * num2)
print("Arithmetic mean:", (num1 + num2) / 2)
print("Larger number:", max(num1, num2))
print("Smaller number:", min(num1, num2))
print("Sum of squares:", num1**2 + num2**2)
```

التمرين 2 – تحليل العددين

المطلوب:

اطلب من المستخدم إدخال عددين صحيحين، ثم احسب واطبع:
1.	حاصل جمعهما.
2.	حاصل ضربهما.
3.	ناتج القسمة العشرية للعدد الأول على الثاني.
4.	ناتج القسمة الصحيحة.
5.	باقي القسمة.
6.	تحقق إذا كان العدد الأول أكبر من العدد الثاني.
7.	تحقق إذا كان العددان متساويين.

الحل:

</div>

<div dir="rtl" style="text-align: right;">  


</div>

```python
num1 = int(input("Enter the first number: "))
num2 = int(input("Enter the second number: "))
print("Sum:", num1 + num2)
print("Multiplication:", num1 * num2)
print("Float division:", num1 / num2)
print("Integer division:", num1 // num2)
print("Remainder:", num1 % num2)
print("Is the first greater than the second?", num1 > num2)
print("Are the two numbers equal?", num1 == num2)
```
