التمرين 1 – أهلية التطعيم


المطلوب:
اطلب من المستخدم إدخال العمر وحالة الصحة ("yes" للتطعيم مسموح، "no" لغير مسموح)، ثم:
1.	إذا كان العمر ≥ 18 وحالته الصحية تسمح، اطبع "مؤهل للتطعيم".
2.	إذا كان العمر < 18 أو حالته الصحية لا تسمح، اطبع "غير مؤهل للتطعيم".
3.	استخدم معامل not للتحقق من الحالة الصحية.
الحل:



</div>

```python
age = int(input("Enter your age: "))
health = input("Is your health condition suitable for vaccination? (yes/no): ")
if age >= 18 and health == "yes":
        print("Eligible for vaccination")
        if age < 18 or health != "yes":
            print("Not eligible for vaccination")
```

التمرين 2 – دخول حفلة

المطلوب:

اطلب من المستخدم إدخال العمر ورخصة الدخول ("yes" أو "no")، ثم:
1.	إذا كان العمر ≥ 21 ولديه رخصة دخول، اطبع "مسموح بالدخول".
2.	إذا كان العمر < 21 أو لا يملك رخصة، اطبع "ممنوع الدخول".
3.	استخدم معامل not للتحقق من الرخصة.
الحل:


</div>

<div dir="rtl" style="text-align: right;">  


</div>

```python
age = int(input("Enter your age: "))
permit = input("Do you have an entry permit? (yes/no): ")
if age >= 21 and permit == "yes":
        print("Entry allowed")
        if age < 21 or permit != "yes":
            print("Entry denied")
```
