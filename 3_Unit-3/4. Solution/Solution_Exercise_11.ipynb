التمرين 1 – تقييم اختبار الطلاب

المطلوب:

اطلب من المستخدم إدخال العمر ودرجة الاختبار (من 100)، ثم:
1.	إذا كان العمر بين 18 و30 والدرجة ≥ 50 → اطبع "ناجح".
2.	إذا كانت الدرجة < 50 → اطبع "راسب".
3.	استخدم جميع معاملات المقارنة على الأقل مرة واحدة.




</div>

```python
age = int(input("Enter your age: "))
score = float(input("Enter your score (out of 100): "))
if age >= 18 and age <= 30 and score >= 50:
    print("Pass")
if score < 50:
        print("Fail")
if age < 18 or age > 30:
        print("Out of evaluation range")
```

التمرين 2 – أهليّة برنامج تدريبي

المطلوب:

اطلب من المستخدم إدخال العمر وعدد ساعات التدريب الأسبوعية، ثم:
1.	إذا كان العمر ≥ 16 و≤ 40 وعدد الساعات ≥ 5 → اطبع "مؤهل للبرنامج".
2.	إذا كانت الساعات < 5 → اطبع "غير مؤهل للبرنامج".
3.	تحقق إذا كان المستخدم خارج الفئة العمرية باستخدام not.
الحل:


</div>

<div dir="rtl" style="text-align: right;">  


</div>

```python
age = int(input("Enter your age: "))
hours = int(input("Number of training hours per week: "))
if age >= 16 and age <= 40 and hours >= 5:
    print("Eligible for the program")
if hours < 5:
    print("Not eligible for the program")
if not (16 <= age <= 40):
    print("Age not suitable for the program")
```
