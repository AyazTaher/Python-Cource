المهمة 1 – قيادة السيارة

المطلوب:

اطلب من المستخدم إدخال العمر وحالة رخصة القيادة ("yes" أو "no")، ثم:
1.	إذا كان العمر ≥ 18 ولديه رخصة، اطبع "مسموح بالقيادة".
2.	إذا كان العمر < 18 أو ليس لديه رخصة، اطبع "غير مسموح بالقيادة".
3.	استخدم معامل not للتحقق من الرخصة.
الحل:



</div>

```python

age = int(input("Enter your age: "))
license_status = input("Do you have a driving license?(yes/no): ")

if age >= 18 and license_status == "yes":
    print("Allowed to drive")
if age < 18 or license_status!= "yes":
        print("Not allowed to drive")
```

المهمة 2 – دخول صالة الألعاب الرياضية

المطلوب:

اطلب من المستخدم إدخال العمر وحالة العضوية ("active" أو "inactive")، ثم:
1.	إذا كان العمر ≥ 16 وحالته نشطة، اطبع "يمكنك الدخول".
2.	إذا كان العمر < 16 أو الحالة غير نشطة، اطبع "لا يمكن الدخول".
3.	استخدم معامل not للتحقق من الحالة.
الحل:


</div>

```python

age = int(input("Enter your age: "))
membership = input("Is your membership active?(active/inactive): ")

if age >= 16 and membership == "active":
    print("Access granted")

if age < 16 or membership!= "active":
        print("Access denied")
```
