المهمة 1 – اختبار اللياقة البدنية

المطلوب:

اطلب من المستخدم إدخال العمر وعدد النقاط في الاختبار (من 100)، ثم:
1.	إذا كان العمر بين 20 و35 والنقاط ≥ 70 → اطبع "ممتاز".
2.	إذا كانت النقاط بين 50 و69 → اطبع "جيد".
3.	إذا كانت النقاط < 50 → اطبع "ضعيف".
4.	استخدم معاملات المقارنة 
الحل:


</div>

```python

age = int(input("Enter your age: "))
score = int(input("Enter your fitness test score (out of 100): "))

if 20 <= age <= 35 and score >= 70:
    print("Excellent")
elif 50 <= score <= 69:
    print("Good")
elif score < 50:
    print("Weak")
else:
    print("Invalid input")
```

المهمة 2 – تقييم المتقدمين لوظيفة

المطلوب:

اطلب من المستخدم إدخال العمر ودرجة المقابلة (من 100)، ثم:
1.	إذا كان العمر ≥ 25 و≤ 35 ودرجة المقابلة ≥ 60 → اطبع "مقبول".
2.	إذا كانت الدرجة < 60 → اطبع "مرفوض".
3.	إذا كان العمر < 25 أو > 35 استخدم not → اطبع "غير مؤهل حسب العمر".
الحل:


</div>

```python

age = int(input("Enter your age: "))
interview_score = int(input("Enter your interview score (out of 100): "))

if 25 <= age <= 35 and interview_score >= 60:
    print("Accepted")
elif interview_score < 60:
    print("Rejected")
elif not (25 <= age <= 35):
    print("Not eligible by age")
else:
    print("Invalid input")



    
```
